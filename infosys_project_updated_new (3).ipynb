**DATA CLEANING -WEEK 2**

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('/content/fifa_players.csv')

# Strip whitespace from column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Drop completely empty rows
df.dropna(how='all', inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Convert date columns
if 'birth_date' in df.columns:
    df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')

# Convert numeric columns
numeric_cols = [
    'age', 'height_cm', 'weight_kgs', 'overall_rating', 'potential',
    'value_euro', 'wage_euro', 'international_reputation(1-5)',
    'weak_foot(1-5)', 'skill_moves(1-5)', 'release_clause_euro',
    'national_rating', 'national_jersey_number', 'crossing', 'finishing',
    'heading_accuracy', 'short_passing', 'volleys', 'dribbling', 'curve',
    'freekick_accuracy', 'long_passing', 'ball_control', 'acceleration',
    'sprint_speed', 'agility', 'reactions', 'balance', 'shot_power',
    'jumping', 'stamina', 'strength', 'long_shots', 'aggression',
    'interceptions', 'positioning', 'vision', 'penalties', 'composure',
    'marking', 'standing_tackle', 'sliding_tackle'
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill missing categorical values
categorical_cols = [
    'name', 'full_name', 'positions', 'nationality', 'preferred_foot',
    'body_type', 'national_team', 'national_team_position'
]

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')

# Optional: Normalize currency columns (e.g., convert to millions)
for col in ['value_euro', 'wage_euro', 'release_clause_euro']:
    if col in df.columns:
        df[col] = df[col] / 1e6  # Convert to millions

#  now Saving cleaned data
df.to_csv('fifa_players_cleaned.csv', index=False)
```

DATA CLEANING FOR GAME EVENTS DATASETS
{dataset has some non-integer numeric values (like NaN or decimals) in the supposed integer ID columns (for example, "player_id", "club_id", etc.).
When we try to convert them to Int64, pandas complains because it can’t safely cast float → int if there are invalid entries.}

```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
#df = pd.read_csv('/content/game_events.csv')
df = pd.read_csv('/content/drive/MyDrive/infosys_project_backup (2)/game_events.csv')

# Step 2: Basic Overview
print("Before cleaning:\n")
print(df.info())
print("\nFirst 5 rows:\n", df.head())

# Step 3: Handle Missing Values
# Fill missing text columns with 'Unknown'
text_cols = ['type', 'description']
for col in text_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')

# Step 4: Convert 'date' column to datetime
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Step 5: Convert ID columns safely
id_cols = ['game_event_id', 'game_id', 'club_id', 'player_id', 'player_in_id', 'player_assist_id']

for col in id_cols:
    if col in df.columns:
        # Convert to numeric, replace non-numeric with NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')
        # Fill NaN with -1, but keep as float to avoid casting issues
        df[col] = df[col].fillna(-1)

# Step 6: Handle minute column
if 'minute' in df.columns:
    df['minute'] = pd.to_numeric(df['minute'], errors='coerce').fillna(0)
    # Ensure realistic range (0–130)
    df.loc[(df['minute'] < 0) | (df['minute'] > 130), 'minute'] = np.nan

# Step 7: Remove duplicates
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"\nRemoved {before - after} duplicate rows.")

# Step 8: Check for remaining missing values
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Step 9: Dataset summary
print("\nSummary after cleaning:")
print(df.describe(include='all'))

# Step 10: Save cleaned file
df.to_csv('game_events_cleaned.csv', index=False)
print("\n Cleaned dataset saved as 'game_events_cleaned.csv'")
```

```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
df = pd.read_csv('/content/drive/MyDrive/infosys_project_backup (2)/dataset.csv')

# Step 2: Quick overview
print("Before Cleaning:\n")
print(df.info())
print("\nFirst 5 rows:\n", df.head())

# Step 3: Convert date column to datetime
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')

# Step 4: Handle missing values

# Fill missing categorical (object) columns
df['position'] = df['position'].fillna('Unknown')

# Fill numeric columns with appropriate strategies
num_cols_mean = ['pace', 'physic']  # few missing, can use mean
for col in num_cols_mean:
    df[col] = df[col].fillna(df[col].mean())

# Columns with large missing data: fill with 0 (represents unavailable previous data)
large_missing_cols = [
    'cumulative_minutes_played', 'cumulative_games_played', 'minutes_per_game_prev_seasons',
    'avg_days_injured_prev_seasons', 'avg_games_per_season_prev_seasons',
    'significant_injury_prev_season', 'cumulative_days_injured',
    'season_days_injured_prev_season'
]
df[large_missing_cols] = df[large_missing_cols].fillna(0)

# Fill missing position_numeric (use median for balance)
df['position_numeric'] = df['position_numeric'].fillna(df['position_numeric'].median())

# Step 5: Handle invalid or inconsistent values

# Ensure no negative numbers in gameplay/injury stats
cols_to_check = [
    'season_days_injured', 'total_days_injured', 'season_minutes_played',
    'season_games_played', 'season_matches_in_squad', 'total_minutes_played',
    'total_games_played', 'age', 'height_cm', 'weight_kg', 'fifa_rating', 'bmi'
]
for col in cols_to_check:
    df[col] = df[col].apply(lambda x: np.nan if x < 0 else x)

# Step 6: Drop duplicates if any
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"\nRemoved {before - after} duplicate rows.")

# Step 7: Recheck missing values
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Step 8: Clean text columns (trim spaces, standardize)
df['nationality'] = df['nationality'].str.strip()
df['work_rate'] = df['work_rate'].str.strip().str.title()
df['position'] = df['position'].str.strip().str.title()

# Step 9: Verify numeric data types
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Step 10: Final dataset summary
print("\nAfter Cleaning:\n")
print(df.describe(include='all'))

# Step 11: Save cleaned dataset
df.to_csv('fifa_players_injury_cleaned.csv', index=False)
print("\n Cleaned dataset saved as 'fifa_players_injury_cleaned.csv'")
```

```python
# --------------------------------
# Step 1: Import Libraries
# --------------------------------
import pandas as pd
import numpy as np
from textblob import TextBlob

# --------------------------------
# Step 2: Load CSVs (No Header Fix)
# --------------------------------
train_path = '/content/drive/MyDrive/infosys_project_backup (2)/twitter_training.csv'
val_path = '/content/drive/MyDrive/infosys_project_backup (2)/twitter_validation.csv'

# These files don’t have headers, so we’ll add our own column names
column_names = ['id', 'entity', 'sentiment', 'text']

train_df = pd.read_csv(train_path, names=column_names, encoding='utf-8')
val_df = pd.read_csv(val_path, names=column_names, encoding='utf-8')

# --------------------------------
# Step 3: Basic Info
# --------------------------------
print("Training Data Info:\n")
print(train_df.info())
print("\nSample Rows:\n")
print(train_df.head())

# --------------------------------
# Step 4: Clean Data
# --------------------------------
train_df.dropna(subset=['text'], inplace=True)
val_df.dropna(subset=['text'], inplace=True)
train_df.drop_duplicates(subset=['text'], inplace=True)
val_df.drop_duplicates(subset=['text'], inplace=True)

# --------------------------------
# Step 5: Sentiment Analysis
# --------------------------------
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

train_df['sentiment_score'] = train_df['text'].apply(get_sentiment)
val_df['sentiment_score'] = val_df['text'].apply(get_sentiment)

# --------------------------------
# Step 6: Label Sentiments
# --------------------------------
def label_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

train_df['sentiment_label'] = train_df['sentiment_score'].apply(label_sentiment)
val_df['sentiment_label'] = val_df['sentiment_score'].apply(label_sentiment)

# --------------------------------
# Step 7: Summary
# --------------------------------
print("\nSentiment Distribution (Training):\n", train_df['sentiment_label'].value_counts())
print("\nSentiment Distribution (Validation):\n", val_df['sentiment_label'].value_counts())

# --------------------------------
# Step 8: Preview Results
# --------------------------------
print("\nSample Output:\n")
print(train_df[['text', 'sentiment_score', 'sentiment_label']].head())
```

```python
import re

def clean_tweet(text):
    text = str(text).lower()
    text = re.sub(r'http\S+', '', text)      # remove URLs
    text = re.sub(r'@\w+', '', text)         # remove mentions
    text = re.sub(r'#', '', text)            # remove hashtags
    text = re.sub(r'[^a-z\s]', '', text)     # remove special characters/numbers
    text = re.sub(r'\s+', ' ', text).strip() # remove extra spaces
    return text

twitter_df['clean_tweet'] = twitter_df['tweet'].apply(clean_tweet)
print("\n Tweets cleaned successfully!")
```

```python
from textblob import TextBlob

def get_polarity(text):
    return TextBlob(text).sentiment.polarity

twitter_df['polarity'] = twitter_df['clean_tweet'].apply(get_polarity)
```

```python
# LSTM Time-series forecasting pipeline

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)



VALUATION_FILE = '/content/drive/MyDrive/infosys_project_backup (2)/player_valuations.csv'   # change if different
FEATURES_FILE = '/content/final_features_for_modeling.csv'  # optional, for multivariate join

assert os.path.exists(VALUATION_FILE), f"Put your valuation file at: {VALUATION_FILE}"
valuation = pd.read_csv(VALUATION_FILE)
print("valuation columns:", valuation.columns.tolist())
# Expect columns like: player_id/full_name, date, market_value_in_eur


#  Preprocess valuation timeseries

# Standardize column names
cols = [c.lower() for c in valuation.columns]
valuation.columns = cols

# Try to find keys
if 'date' not in valuation.columns:
    # try common synonyms
    for cand in ['timestamp','time','dt']:
        if cand in valuation.columns:
            valuation.rename(columns={cand:'date'}, inplace=True)
            break

if 'market_value_in_eur' not in valuation.columns and 'value' in valuation.columns:
    valuation.rename(columns={'value':'market_value_in_eur'}, inplace=True)

# Find player identifier
player_col = None
for cand in ['player_id','player','full_name','name']:
    if cand in valuation.columns:
        player_col = cand
        break
if player_col is None:
    raise KeyError("Can't find a player id/name column in valuation file. Columns: " + str(valuation.columns.tolist()))

valuation['date'] = pd.to_datetime(valuation['date'], errors='coerce')
valuation = valuation.dropna(subset=['date', 'market_value_in_eur'])
valuation = valuation.sort_values([player_col, 'date']).reset_index(drop=True)

# Optional: inspect a single player's time series length distribution
counts = valuation.groupby(player_col).size().sort_values(ascending=False)
print("Top players by number of valuation points:\n", counts.head())

# For demonstration, choose one player with enough history
min_points = 30
candidates = counts[counts >= min_points].index.tolist()
if len(candidates) == 0:
    print("No player has >= {} points. Lower min_points or create a synthetic example.".format(min_points))
    sample_player = counts.index[0]
else:
    sample_player = candidates[0]
print("Using sample player:", sample_player)

player_ts = valuation[valuation[player_col] == sample_player].copy()
player_ts = player_ts.set_index('date').asfreq('MS')  # monthly frequency (change if your data is weekly/daily)
# forward/backfill missing market values
player_ts['market_value_in_eur'] = player_ts['market_value_in_eur'].interpolate().fillna(method='bfill').fillna(method='ffill')

# If there's still NaN, drop them
player_ts = player_ts.dropna(subset=['market_value_in_eur'])
print("Series length for sample player:", len(player_ts))


# Helper: sliding windows

def make_windows(series, lookback=12, horizon=1):
    """
    Inputs:
      series: 1D numpy array or pandas Series
      lookback: number of past timesteps (X length)
      horizon: number of future timesteps to predict (y length)
    Returns:
      X shape: (n_samples, lookback, 1)
      y shape: (n_samples, horizon)
    """
    s = np.array(series).astype(float)
    X, y = [], []
    for i in range(len(s) - lookback - horizon + 1):
        X.append(s[i : i + lookback])
        y.append(s[i + lookback : i + lookback + horizon])
    X = np.array(X)
    y = np.array(y)
    # reshape X to (samples, lookback, 1)
    return X.reshape((X.shape[0], X.shape[1], 1)), y


#  Univariate LSTM (one-step)

lookback = 12   # use 12 months of history
horizon = 1     # one-step ahead

series = player_ts['market_value_in_eur'].values
scaler_uni = MinMaxScaler()
series_scaled = scaler_uni.fit_transform(series.reshape(-1,1)).flatten()

X, y = make_windows(series_scaled, lookback=lookback, horizon=horizon)
print("X,y shapes (univariate):", X.shape, y.shape)

# train/test split (time-series aware)
train_frac = 0.8
n_train = int(len(X) * train_frac)
X_train, X_val = X[:n_train], X[n_train:]
y_train, y_val = y[:n_train], y[n_train:]

# Build LSTM model
def build_univariate_lstm(input_shape, units=50, dropout=0.1):
    model = Sequential([
        LSTM(units, input_shape=input_shape),
        Dropout(dropout),
        Dense(horizon)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

uni_model = build_univariate_lstm((lookback, 1), units=64, dropout=0.2)
uni_model.summary()

# Callbacks
es = EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True)
ckp_path = "/content/univariate_lstm.h5"
mck = ModelCheckpoint(ckp_path, save_best_only=True, monitor='val_loss')

history_uni = uni_model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=200,
    batch_size=16,
    callbacks=[es, mck],
    verbose=2
)

# Evaluate and plot loss
plt.figure(figsize=(6,4))
plt.plot(history_uni.history['loss'], label='train')
plt.plot(history_uni.history['val_loss'], label='val')
plt.title('Univariate LSTM Loss')
plt.xlabel('epoch'); plt.ylabel('mse'); plt.legend(); plt.show()

# Predictions & metrics (inverse scaling)
y_pred_val = uni_model.predict(X_val)
# y_val shape: (n_samples, 1)
y_val_inv = scaler_uni.inverse_transform(y_val.reshape(-1,1)).flatten()
y_pred_inv = scaler_uni.inverse_transform(y_pred_val.reshape(-1,1)).flatten()

import numpy as np

mae = mean_absolute_error(y_val_inv, y_pred_inv)
mse = mean_squared_error(y_val_inv, y_pred_inv)
rmse = np.sqrt(mse)
print("Univariate LSTM — Val MAE: {:.2f}, RMSE: {:.2f}".format(mae, rmse))

# Plot actual vs predicted for validation window
plt.figure(figsize=(8,4))
plt.plot(y_val_inv, label='actual')
plt.plot(y_pred_inv, label='predicted')
plt.title('Univariate LSTM — actual vs predicted (validation)')
plt.legend(); plt.show()


#  Multivariate LSTM (injury + sentiment)

# For multivariate we need features per date. If you have a features timeseries (per player per date),
# join them to valuation by date. If not available, we demonstrate creating simple synthetic features:
# - avg_days_injured: slowly varying series
# - avg_sentiment: synthetic or merged from sentiment per date (if available)

# Attempt to load a features-by-date file (optional)
if os.path.exists(FEATURES_FILE):
    feats = pd.read_csv(FEATURES_FILE)
    # try to align by date/player; this part depends strongly on how your features are stored.
    # For simplicity, try to find columns: 'full_name','date','avg_sentiment','avg_days_injured'
    feats.columns = [c.lower() for c in feats.columns]
    if 'date' in feats.columns:
        feats['date'] = pd.to_datetime(feats['date'], errors='coerce')
        # filter for the sample player and join
        if 'full_name' in feats.columns:
            p_feats = feats[feats['full_name'] == sample_player].set_index('date').sort_index()
            # reindex to player_ts index and forward fill
            p_feats = p_feats.reindex(player_ts.index).interpolate().fillna(method='bfill').fillna(method='ffill')
            # choose columns if present else make synthetic
            if 'avg_sentiment' in p_feats.columns:
                sentiment_series = p_feats['avg_sentiment'].values
            else:
                sentiment_series = np.zeros(len(player_ts))
            if 'avg_days_injured' in p_feats.columns:
                injury_series = p_feats['avg_days_injured'].values
            else:
                injury_series = np.zeros(len(player_ts))
        else:
            # fallback synthetic
            sentiment_series = np.zeros(len(player_ts))
            injury_series = np.zeros(len(player_ts))
    else:
        sentiment_series = np.zeros(len(player_ts))
        injury_series = np.zeros(len(player_ts))
else:
    # If no features file, create two synthetic features to demonstrate multivariate model
    n = len(player_ts)
    np.random.seed(SEED)
    sentiment_series = (np.sin(np.linspace(0,3.0,n)) * 0.2) + 0.05*np.random.randn(n)  # synthetic
    injury_series = (np.clip(np.cos(np.linspace(0,2.0,n))*0.1 + 0.05*np.random.randn(n), 0, None))

# Build multivariate dataset: columns = [market_value, sentiment, injury]
df_multi = pd.DataFrame({
    'market_value': player_ts['market_value_in_eur'].values,
    'sentiment': sentiment_series,
    'injury': injury_series
}, index=player_ts.index)

# Normalize features (fit scaler on train portion to avoid leakage)
lookback = 12
horizon = 1
values = df_multi.values  # shape (T, 3)
scaler_multi = MinMaxScaler()
values_scaled = scaler_multi.fit_transform(values)

# create windows (multivariate)
def make_windows_multi(arr, lookback=12, horizon=1):
    X, y = [], []
    for i in range(len(arr) - lookback - horizon + 1):
        X.append(arr[i:i+lookback, :])      # (lookback, n_features)
        y.append(arr[i+lookback:i+lookback+horizon, 0])  # target is column 0 (market_value)
    return np.array(X), np.array(y)

X_m, y_m = make_windows_multi(values_scaled, lookback=lookback, horizon=horizon)
print("Multivariate X,y shapes:", X_m.shape, y_m.shape)

# train/val split
n_train = int(len(X_m) * 0.8)
Xtr, Xv = X_m[:n_train], X_m[n_train:]
ytr, yv = y_m[:n_train], y_m[n_train:]

# build multivariate LSTM
def build_multivariate_lstm(input_shape, units=64, dropout=0.2):
    model = Sequential()
    model.add(LSTM(units, input_shape=input_shape))
    model.add(Dropout(dropout))
    model.add(Dense(horizon))
    model.compile(optimizer='adam', loss='mse')
    return model

mv_model = build_multivariate_lstm((lookback, X_m.shape[2]), units=64, dropout=0.2)
mv_model.summary()

es2 = EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True)
ckp2 = ModelCheckpoint('/content/multivariate_lstm.h5', save_best_only=True, monitor='val_loss')

hist_mv = mv_model.fit(Xtr, ytr, validation_data=(Xv, yv), epochs=200, batch_size=16, callbacks=[es2, ckp2], verbose=2)

# plot loss
plt.figure(figsize=(6,4))
plt.plot(hist_mv.history['loss'], label='train')
plt.plot(hist_mv.history['val_loss'], label='val')
plt.title('Multivariate LSTM Loss'); plt.legend(); plt.show()

# Evaluate multivariate model
yv_pred = mv_model.predict(Xv)
# yv and yv_pred are scaled target values; inverse transform requires special handling:
# We only scaled full multivariate arrays; to inverse the target we create dummy arrays
def inv_target(scaled_targets):
    # scaled_targets shape: (n_samples, 1)
    # create dummy array with same shape as original features, replace first column
    n = scaled_targets.shape[0]
    dummy = np.zeros((n, values.shape[1]))
    dummy[:,0] = scaled_targets.flatten()
    inv = scaler_multi.inverse_transform(dummy)[:,0]
    return inv

yv_inv = inv_target(yv)
yv_pred_inv = inv_target(yv_pred)

import numpy as np

mae_mv = mean_absolute_error(yv_inv, yv_pred_inv)
mse_mv = mean_squared_error(yv_inv, yv_pred_inv)
rmse_mv = np.sqrt(mse_mv)
print("Multivariate LSTM — Val MAE: {:.2f}, RMSE: {:.2f}".format(mae_mv, rmse_mv))


plt.figure(figsize=(8,4))
plt.plot(yv_inv, label='actual')
plt.plot(yv_pred_inv, label='predicted')
plt.title('Multivariate LSTM — actual vs predicted (validation)')
plt.legend(); plt.show()


# Encoder-Decoder LSTM for multi-step forecasting

# We will predict multi-step horizon (e.g., next 3 periods)
lookback = 12
horizon = 3   # predict next 3 months
# prepare data for encoder-decoder: X (lookback, n_features), Y (horizon,)
X_enc, y_enc = make_windows_multi(values_scaled, lookback=lookback, horizon=horizon)
print("Encoder-decoder dataset shapes:", X_enc.shape, y_enc.shape)

n_train = int(len(X_enc) * 0.8)
Xtr_e, Xv_e = X_enc[:n_train], X_enc[n_train:]
ytr_e, yv_e = y_enc[:n_train], y_enc[n_train:]

n_features = X_enc.shape[2]

# Encoder-Decoder model
latent_dim = 64

encoder_inputs = Input(shape=(lookback, n_features))
encoder = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_inputs)
encoder_states = [state_h, state_c]

decoder_inputs = RepeatVector(horizon)(encoder_outputs)  # shape -> (batch, horizon, latent_dim)
decoder_lstm = LSTM(latent_dim, return_sequences=True)
decoder_outputs = decoder_lstm(decoder_inputs, initial_state=encoder_states)
decoder_dense = TimeDistributed(Dense(1))  # predict single value at each future step
decoder_outputs = decoder_dense(decoder_outputs)

encdec_model = Model(encoder_inputs, decoder_outputs)
encdec_model.compile(optimizer='adam', loss='mse')
encdec_model.summary()

es3 = EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True)
ckp3 = ModelCheckpoint('/content/encdec_lstm.h5', save_best_only=True, monitor='val_loss')

# ytr_e shape is (n_samples, horizon) but decoder expects (n_samples, horizon, 1)
ytr_e3 = ytr_e.reshape((ytr_e.shape[0], ytr_e.shape[1], 1))
yv_e3 = yv_e.reshape((yv_e.shape[0], yv_e.shape[1], 1))

hist_encdec = encdec_model.fit(
    Xtr_e, ytr_e3,
    validation_data=(Xv_e, yv_e3),
    epochs=200, batch_size=16,
    callbacks=[es3, ckp3], verbose=2
)

# Plot loss
plt.figure(figsize=(6,4))
plt.plot(hist_encdec.history['loss'], label='train')
plt.plot(hist_encdec.history['val_loss'], label='val')
plt.title('Encoder-Decoder LSTM Loss'); plt.legend(); plt.show()

# Predict and invert scale for multi-step
yv_pred_enc = encdec_model.predict(Xv_e)  # shape (n_samples, horizon, 1)
# invert each horizon step
def inv_multi_step(pred_array):
    # pred_array shape: (n_samples, horizon, 1)
    n_samples, h, _ = pred_array.shape
    inv = np.zeros((n_samples, h))
    for t in range(h):
        col = pred_array[:, t, 0]
        dummy = np.zeros((len(col), values.shape[1]))
        dummy[:,0] = col
        inv[:,t] = scaler_multi.inverse_transform(dummy)[:,0]
    return inv

yv_enc_inv = inv_multi_step(yv_e3)
yv_pred_enc_inv = inv_multi_step(yv_pred_enc)

# compute metrics per horizon step and overall (flatten)
mae_enc = mean_absolute_error(yv_enc_inv.flatten(), yv_pred_enc_inv.flatten())
rmse_enc = mean_squared_error(yv_enc_inv.flatten(), yv_pred_enc_inv.flatten(), squared=False)
print("Encoder-Decoder LSTM — Val MAE (multi-step): {:.2f}, RMSE: {:.2f}".format(mae_enc, rmse_enc))

# Plot first horizon actual vs predicted for validation
plt.figure(figsize=(8,4))
plt.plot(yv_enc_inv[:,0], label='actual step1')
plt.plot(yv_pred_enc_inv[:,0], label='predicted step1')
plt.title('Encoder-Decoder — step 1 actual vs predicted')
plt.legend(); plt.show()


#  Save models and final notes

uni_model.save('/content/univariate_lstm_final.h5')
mv_model.save('/content/multivariate_lstm_final.h5')
encdec_model.save('/content/encdec_lstm_final.h5')
print("Saved models to /content/*.h5")
```

```python

# LSTM Model Development for Player Value Prediction



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, Input, RepeatVector, TimeDistributed


# 2. Load and Prepare Data

df = pd.read_csv("final_features_for_modeling.csv")

# Ensure proper columns exist
df = df.dropna(subset=['value_euro'])

# Sort players for consistency
df = df.sort_values(by=['age']).reset_index(drop=True)

# Normalize for LSTM
scaler = MinMaxScaler()
df['scaled_value'] = scaler.fit_transform(df[['value_euro']])


# 3. Helper Function to Create Sequences

def create_sequences(data, time_steps=5):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i+time_steps])
        y.append(data[i+time_steps])
    return np.array(X), np.array(y)


# UNIVARIATE LSTM (Predicting Market Value only)

values = df['scaled_value'].values
time_steps = 5
X, y = create_sequences(values, time_steps)

# Split train/val
train_size = int(0.8 * len(X))
X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:], y[train_size:]

# Reshape
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))

# Build Model
model_uni = Sequential([
    LSTM(64, activation='tanh', input_shape=(time_steps, 1)),
    Dense(1)
])
model_uni.compile(optimizer='adam', loss='mse')
history_uni = model_uni.fit(X_train, y_train, epochs=20, batch_size=16,
                            validation_data=(X_val, y_val), verbose=1)

# Predictions
y_pred = model_uni.predict(X_val)

# Inverse scaling
y_val_inv = scaler.inverse_transform(y_val.reshape(-1, 1))
y_pred_inv = scaler.inverse_transform(y_pred)

# Evaluation
mae = mean_absolute_error(y_val_inv, y_pred_inv)
mse = mean_squared_error(y_val_inv, y_pred_inv)
rmse = np.sqrt(mse)
print("Univariate LSTM — Val MAE: {:.2f}, RMSE: {:.2f}".format(mae, rmse))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(y_val_inv, label='Actual')
plt.plot(y_pred_inv, label='Predicted')
plt.title("Univariate LSTM — Actual vs Predicted Value")
plt.xlabel("Samples")
plt.ylabel("Market Value (Euros)")
plt.legend()
plt.show()


# MULTIVARIATE LSTM (Using Additional Features)

features = ['overall_rating', 'potential', 'avg_days_injured',
            'injury_value_impact', 'avg_sentiment', 'scaled_value']

data_mv = df[features].fillna(0)
scaler_mv = MinMaxScaler()
data_mv_scaled = scaler_mv.fit_transform(data_mv)

# Split input-output
X_mv, y_mv = create_sequences(data_mv_scaled, time_steps)
y_mv = y_mv[:, -1]  # last value_euro step

train_size = int(0.8 * len(X_mv))
X_train_mv, y_train_mv = X_mv[:train_size], y_mv[:train_size]
X_val_mv, y_val_mv = X_mv[train_size:], y_mv[train_size:]

# Build Model
model_mv = Sequential([
    LSTM(64, activation='tanh', input_shape=(time_steps, X_mv.shape[2])),
    Dense(1)
])
model_mv.compile(optimizer='adam', loss='mse')
history_mv = model_mv.fit(X_train_mv, y_train_mv, epochs=20, batch_size=16,
                          validation_data=(X_val_mv, y_val_mv), verbose=1)

# Predictions
y_pred_mv = model_mv.predict(X_val_mv)

# Inverse scaling (only for target)
target_scaler = MinMaxScaler()
target_scaler.min_, target_scaler.scale_ = scaler.data_min_[0], scaler.scale_[0]
yv_inv = scaler.inverse_transform(y_val_mv.reshape(-1, 1))
yv_pred_inv = scaler.inverse_transform(y_pred_mv)

# Evaluation
mae_mv = mean_absolute_error(yv_inv, yv_pred_inv)
mse_mv = mean_squared_error(yv_inv, yv_pred_inv)
rmse_mv = np.sqrt(mse_mv)
print("Multivariate LSTM — Val MAE: {:.2f}, RMSE: {:.2f}".format(mae_mv, rmse_mv))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(yv_inv, label='Actual')
plt.plot(yv_pred_inv, label='Predicted')
plt.title("Multivariate LSTM — Actual vs Predicted Value")
plt.xlabel("Samples")
plt.ylabel("Market Value (Euros)")
plt.legend()
plt.show()


# ENCODER-DECODER LSTM (Multi-step Forecasting)

forecast_steps = 3  # predict next 3 transfer windows

def create_multistep_sequences(data, time_steps=5, forecast_steps=3):
    X, y = [], []
    for i in range(len(data) - time_steps - forecast_steps):
        X.append(data[i:i+time_steps])
        y.append(data[i+time_steps:i+time_steps+forecast_steps])
    return np.array(X), np.array(y)

X_enc, y_enc = create_multistep_sequences(values, time_steps, forecast_steps)

train_size = int(0.8 * len(X_enc))
X_train_enc, y_train_enc = X_enc[:train_size], y_enc[:train_size]
X_val_enc, y_val_enc = X_enc[train_size:], y_enc[train_size:]

# Reshape
X_train_enc = X_train_enc.reshape((X_train_enc.shape[0], X_train_enc.shape[1], 1))
y_train_enc = y_train_enc.reshape((y_train_enc.shape[0], y_train_enc.shape[1], 1))
X_val_enc = X_val_enc.reshape((X_val_enc.shape[0], X_val_enc.shape[1], 1))
y_val_enc = y_val_enc.reshape((y_val_enc.shape[0], y_val_enc.shape[1], 1))

# Build Encoder-Decoder Model
enc_inputs = Input(shape=(time_steps, 1))
encoded = LSTM(64, activation='tanh')(enc_inputs)
rep = RepeatVector(forecast_steps)(encoded)
dec_outputs = LSTM(64, activation='tanh', return_sequences=True)(rep)
outputs = TimeDistributed(Dense(1))(dec_outputs)
model_encdec = Model(enc_inputs, outputs)
model_encdec.compile(optimizer='adam', loss='mse')
history_encdec = model_encdec.fit(X_train_enc, y_train_enc, epochs=20, batch_size=16,
                                  validation_data=(X_val_enc, y_val_enc), verbose=1)

# Predictions
y_pred_enc = model_encdec.predict(X_val_enc)

# Inverse scale
yv_enc_inv = scaler.inverse_transform(y_val_enc.reshape(-1, 1))
yv_pred_enc_inv = scaler.inverse_transform(y_pred_enc.reshape(-1, 1))

# Evaluation
mae_enc = mean_absolute_error(yv_enc_inv.flatten(), yv_pred_enc_inv.flatten())
mse_enc = mean_squared_error(yv_enc_inv.flatten(), yv_pred_enc_inv.flatten())
rmse_enc = np.sqrt(mse_enc)
print("Encoder-Decoder LSTM — Val MAE (multi-step): {:.2f}, RMSE: {:.2f}".format(mae_enc, rmse_enc))

# Plot Multi-step Prediction
plt.figure(figsize=(8, 5))
plt.plot(yv_enc_inv[:100], label="Actual")
plt.plot(yv_pred_enc_inv[:100], label="Predicted")
plt.title("Encoder-Decoder LSTM — Multi-step Forecasting")
plt.xlabel("Time Steps")
plt.ylabel("Market Value (Euros)")
plt.legend()
plt.show()


# PERFORMANCE VISUALIZATION

plt.figure(figsize=(10, 4))
plt.plot(history_uni.history['loss'], label='Univariate Train Loss')
plt.plot(history_uni.history['val_loss'], label='Univariate Val Loss')
plt.title("Univariate LSTM Loss")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(history_mv.history['loss'], label='Multivariate Train Loss')
plt.plot(history_mv.history['val_loss'], label='Multivariate Val Loss')
plt.title("Multivariate LSTM Loss")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(history_encdec.history['loss'], label='Encoder-Decoder Train Loss')
plt.plot(history_encdec.history['val_loss'], label='Encoder-Decoder Val Loss')
plt.title("Encoder-Decoder LSTM Loss")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.show()

print("\n🎯 All LSTM Models Trained & Evaluated Successfully!")
```

```python
# Milestone 6 & 7 — Full Code
# Requirements: tensorflow, keras_tuner, scikit-learn, xgboost, joblib, plotly, pandas, numpy
# Make sure you have X_train, X_test, y_train, y_test (and optionally X_holdout, y_holdout) defined.

# -----------------------
# Imports & Utility funcs
# -----------------------
import os
import numpy as np
import pandas as pd

# ML / DL
import joblib
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import xgboost as xgb

# Plotting
import plotly.express as px
import plotly.graph_objects as go

# Keras/TensorFlow & Tuner
import tensorflow as tf
from tensorflow import keras

try:
    import keras_tuner as kt
except Exception:
    # fallback import name used by some installs
    import kerastuner as kt

# -----------------------
# Basic environment checks
# -----------------------
def ensure_vars_exist(var_names, local_vars):
    missing = [v for v in var_names if v not in local_vars]
    if missing:
        raise RuntimeError(f"Missing required variables in the environment: {missing}. "
                           "Define them before running this script (e.g., X_train, y_train, X_test, y_test).")

# call this at start of main functions

# -----------------------
# 1) LSTM Hyperparameter Tuning (Keras Tuner)
# -----------------------
def build_lstm_model(hp, input_shape):
    """
    Keras Tuner-compatible builder for an LSTM regression model.
    hp: hyperparameter object provided by Keras Tuner
    input_shape: tuple like (timesteps, features) OR (features,) for dense input
    """
    model = keras.Sequential()

    # Tune whether to use LSTM (for sequence data) or dense for tabular
    # For tabular data we wrap in a Dense layer, for sequence shape use LSTM
    hp_units = hp.Int('units', min_value=32, max_value=256, step=32)
    hp_dropout = hp.Float('dropout', 0.0, 0.5, step=0.1)
    hp_lr = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])

    # If input_shape has length 2, assume sequence (timesteps, features)
    if len(input_shape) == 2:
        model.add(keras.layers.Input(shape=input_shape))
        model.add(keras.layers.LSTM(hp_units, return_sequences=False))
    else:
        model.add(keras.layers.Input(shape=(input_shape[0],)))
        model.add(keras.layers.Dense(hp_units, activation='relu'))

    model.add(keras.layers.Dropout(hp_dropout))
    model.add(keras.layers.Dense(1, activation='linear'))

    model.compile(optimizer=keras.optimizers.Adam(learning_rate=hp_lr),
                  loss='mse',
                  metrics=['mae'])
    return model

def tune_lstm(X_train, y_train, input_shape=None, max_trials=10, epochs=20, project_dir='lstm_tuning'):
    """
    Run Keras Tuner RandomSearch to tune LSTM model.
    X_train, y_train: training arrays
    input_shape: if None, derived from X_train shape (works for both seq and tabular)
    Returns best model (compiled).
    """
    ensure_vars_exist(['X_train','y_train'], locals())
    if input_shape is None:
        # Derive input shape: if 3D, assume (samples, timesteps, features)
        shape = X_train.shape[1:]
    else:
        shape = input_shape

    def builder(hp):
        return build_lstm_model(hp, input_shape=shape)

    tuner = kt.RandomSearch(
        builder,
        objective='val_mae',
        max_trials=max_trials,
        overwrite=True,
        directory=project_dir,
        project_name='player_value_lstm'
    )

    tuner.search(X_train, y_train, epochs=epochs, validation_split=0.15, verbose=1)
    best_models = tuner.get_best_models(num_models=1)
    if len(best_models) == 0:
        raise RuntimeError("Tuner returned no best models. Check tuner logs.")
    best_model = best_models[0]
    return best_model, tuner

# -----------------------
# 2) Random Forest & XGBoost Hyperparameter Search
# -----------------------
def tune_random_forest(X_train, y_train, cv=3, scoring='neg_mean_absolute_error'):
    rf_grid = {
        "n_estimators": [100, 200, 400],
        "max_depth": [10, 20, 30, None],
        "min_samples_split": [2, 5, 10]
    }
    rf = RandomForestRegressor(random_state=42, n_jobs=-1)
    rf_search = GridSearchCV(rf, rf_grid, cv=cv, scoring=scoring, n_jobs=-1)
    rf_search.fit(X_train, y_train)
    return rf_search.best_estimator_, rf_search

def tune_xgboost(X_train, y_train, cv=3, scoring='neg_mean_absolute_error'):
    params = {
        "max_depth": [3, 5, 7],
        "learning_rate": [0.05, 0.1, 0.2],
        "n_estimators": [100, 200, 400],
        "subsample": [0.7, 1.0]
    }
    xgb_model = xgb.XGBRegressor(objective='reg:squarederror', n_jobs=-1, random_state=42)
    xgb_search = GridSearchCV(xgb_model, params, cv=cv, scoring=scoring, n_jobs=-1)
    xgb_search.fit(X_train, y_train)
    return xgb_search.best_estimator_, xgb_search

# -----------------------
# 3) Model Evaluation Helpers
# -----------------------
def evaluate_model(model, X, y_true, name="model"):
    """
    Generic evaluate: returns dict with rmse & mae
    Works with sklearn-like .predict and Keras models (wrapped)
    """
    # Keras models sometimes output shape (n,1) => flatten
    preds = model.predict(X)
    preds = np.array(preds).reshape(-1)
    y_true = np.array(y_true).reshape(-1)
    rmse = np.sqrt(mean_squared_error(y_true, preds))
    mae = mean_absolute_error(y_true, preds)
    return {"rmse": float(rmse), "mae": float(mae), "preds": preds}

# -----------------------
# 4) Save & Load Models
# -----------------------
def save_models(models_dict, save_dir='saved_models'):
    os.makedirs(save_dir, exist_ok=True)
    saved_paths = {}
    for name, model in models_dict.items():
        if hasattr(model, 'save') and callable(model.save) and name.lower().find('lstm')!=-1:
            # Keras model
            path = os.path.join(save_dir, f"{name.replace(' ', '_')}.h5")
            model.save(path)
            saved_paths[name] = path
        else:
            # joblib for sklearn/xgboost
            path = os.path.join(save_dir, f"{name.replace(' ', '_')}.pkl")
            joblib.dump(model, path)
            saved_paths[name] = path
    return saved_paths

def load_model_by_path(path):
    if path.endswith('.h5'):
        return keras.models.load_model(path)
    elif path.endswith('.pkl'):
        return joblib.load(path)
    else:
        raise ValueError("Unknown model file extension (supported: .h5, .pkl)")

# -----------------------
# 5) Visualization Helpers
# -----------------------
def plot_actual_vs_pred(y_true, preds, title="Actual vs Predicted", show_plot=True):
    df = pd.DataFrame({"actual": np.array(y_true).reshape(-1), "pred": np.array(preds).reshape(-1)})
    fig = px.line(df, y=['actual','pred'], title=title)
    if show_plot:
        fig.show()
    return fig

def plot_feature_importances(feature_names, importances, title="Feature Importances", show_plot=True):
    df = pd.DataFrame({"feature": feature_names, "importance": importances})
    df = df.sort_values("importance", ascending=True)
    fig = px.bar(df, x='importance', y='feature', orientation='h', title=title)
    if show_plot:
        fig.show()
    return fig

# -----------------------
# 6) Deployment Helper
# -----------------------
def predict_with_model(model, X):
    """
    General predict wrapper (handles sklearn, xgboost, keras)
    """
    preds = model.predict(X)
    return np.array(preds).reshape(-1)

def create_simple_predict_function(model):
    """
    Returns a function that accepts a DataFrame or array and returns predictions.
    Useful for quick local deployment or integration into an API.
    """
    def predict_fn(X_input):
        # If DataFrame, convert to numpy
        if isinstance(X_input, pd.DataFrame):
            arr = X_input.values
        else:
            arr = np.array(X_input)
        return predict_with_model(model, arr)
    return predict_fn

# -----------------------
# 7) Full Orchestration (run all steps)
# -----------------------
def run_full_pipeline(X_train, y_train, X_test, y_test,
                      X_holdout=None, y_holdout=None,
                      do_tune_lstm=True, do_tune_rf=True, do_tune_xgb=True,
                      lstm_trials=6, lstm_epochs=20):
    """
    Orchestrate tuning, selection, validation, saving, and visualization.
    Returns a dict with results and saved model paths.
    """
    # Basic checks
    ensure_vars_exist(['X_train','y_train','X_test','y_test'], locals())
    results = {}
    models = {}

    # LSTM tuning (only if data shape suits it)
    try:
        if do_tune_lstm:
            print("Tuning LSTM (this can take a while)...")
            best_lstm, tuner = tune_lstm(X_train, y_train, max_trials=lstm_trials, epochs=lstm_epochs)
            models['LSTM'] = best_lstm
            results['LSTM_tuner_summary'] = tuner.results_summary() if hasattr(tuner, 'results_summary') else "tuner run"
            print("LSTM tuning complete.")
    except Exception as e:
        print("LSTM tuning skipped/failed:", e)

    # Random Forest
    try:
        if do_tune_rf:
            print("Tuning Random Forest (GridSearch)...")
            best_rf, rf_search = tune_random_forest(X_train, y_train)
            models['RandomForest'] = best_rf
            results['RandomForest_cv_results'] = rf_search.best_params_
            print("Random Forest tuning complete. Best params:", rf_search.best_params_)
    except Exception as e:
        print("Random Forest tuning skipped/failed:", e)

    # XGBoost
    try:
        if do_tune_xgb:
            print("Tuning XGBoost (GridSearch)...")
            best_xgb, xgb_search = tune_xgboost(X_train, y_train)
            models['XGBoost'] = best_xgb
            results['XGBoost_cv_results'] = xgb_search.best_params_
            print("XGBoost tuning complete. Best params:", xgb_search.best_params_)
    except Exception as e:
        print("XGBoost tuning skipped/failed:", e)

    # Evaluate all models on test set
    eval_table = {}
    for name, model in models.items():
        try:
            ev = evaluate_model(model, X_test, y_test, name)
            eval_table[name] = {"rmse": ev['rmse'], "mae": ev['mae']}
            # keep preds for plotting if needed
            results[f"{name}_preds_test"] = ev['preds']
            print(f"{name} evaluation on test set -> RMSE: {ev['rmse']:.4f}, MAE: {ev['mae']:.4f}")
        except Exception as e:
            print(f"Evaluation failed for {name}:", e)

    # If holdout provided, evaluate on holdout
    if X_holdout is not None and y_holdout is not None:
        holdout_results = {}
        for name, model in models.items():
            try:
                hev = evaluate_model(model, X_holdout, y_holdout, name)
                holdout_results[name] = {"rmse": hev['rmse'], "mae": hev['mae']}
                print(f"{name} evaluation on HOLDOUT -> RMSE: {hev['rmse']:.4f}, MAE: {hev['mae']:.4f}")
            except Exception as e:
                print(f"Holdout evaluation failed for {name}:", e)
        results['holdout_eval'] = holdout_results

    results['test_eval'] = eval_table

    # Choose best model by RMSE then MAE (simple strategy)
    best_name = None
    best_rmse = np.inf
    for name, metrics in eval_table.items():
        if metrics['rmse'] < best_rmse:
            best_rmse = metrics['rmse']
            best_name = name
    print(f"Selected final model: {best_name} (by RMSE)")

    # Save models
    saved_paths = save_models(models, save_dir='saved_models')
    print("Models saved to:", saved_paths)

    results['models'] = models
    results['saved_paths'] = saved_paths
    results['selected_model_name'] = best_name
    results['selected_model'] = models.get(best_name, None)

    # Visualizations: Actual vs Pred for selected model
    if results['selected_model'] is not None:
        sel_preds = results[f"{best_name}_preds_test"]
        plot_actual_vs_pred(y_test, sel_preds, title=f"Actual vs Predicted ({best_name})")

    # Feature importance plot if RandomForest present and X_train is DataFrame
    if 'RandomForest' in models and isinstance(X_train, pd.DataFrame):
        try:
            fi = models['RandomForest'].feature_importances_
            plot_feature_importances(X_train.columns.tolist(), fi, title="RF Feature Importances")
        except Exception as e:
            print("Could not plot feature importances:", e)

    return results

# -----------------------
# 8) Example usage
# -----------------------
# NOTE: replace the following calls with your actual variables in the notebook.
# Example:
# results = run_full_pipeline(X_train, y_train, X_test, y_test, X_holdout=X_holdout, y_holdout=y_holdout,
#                             do_tune_lstm=True, do_tune_rf=True, do_tune_xgb=True,
#                             lstm_trials=6, lstm_epochs=20)

# This will:
# - Run LSTM tuning (if enabled)
# - Run RF and XGB GridSearch
# - Evaluate models on test (+ holdout if supplied)
# - Save models to ./saved_models
# - Display interactive Plotly charts in the notebook

# -----------------------
# 9) Quick Predict API example (Flask skeleton)
# -----------------------
flask_api_example = """
# Minimal Flask app skeleton for serving the selected model.
# Save this as serve_model.py and run: FLASK_APP=serve_model.py flask run

from flask import Flask, request, jsonify
import joblib
import numpy as np
import pandas as pd
from tensorflow import keras

app = Flask(__name__)
# Load your selected model
model = joblib.load('saved_models/RandomForest.pkl')  # or keras.models.load_model('saved_models/LSTM.h5')

@app.route('/predict', methods=['POST'])
def predict():
    data = request.json
    # data expected: {"instances": [[f1, f2, ...], [...]]}
    instances = data.get('instances')
    arr = np.array(instances)
    preds = model.predict(arr).reshape(-1).tolist()
    return jsonify({'predictions': preds})

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
"""

# -----------------------
# End of script
# -----------------------
```

```python
# Milestone 6 & 7 — Full Code (NO KERAS TUNER)
# This version is fully compatible with your current Colab environment.

# -----------------------
# Imports
# -----------------------
import os
import numpy as np
import pandas as pd

import joblib
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import xgboost as xgb

import plotly.express as px
import plotly.graph_objects as go

import tensorflow as tf
from tensorflow import keras


# -----------------------
# Utility checks
# -----------------------
def ensure_vars_exist(var_names, local_vars):
    missing = [v for v in var_names if v not in local_vars]
    if missing:
        raise RuntimeError(f"Missing variables: {missing}")


# -----------------------
# 1) Manual LSTM Hyperparameter Search (no keras-tuner)
# -----------------------
def build_lstm(units=64, dropout=0.2, lr=0.001, input_shape=None):
    model = keras.Sequential()
    model.add(keras.layers.LSTM(units, input_shape=input_shape))
    model.add(keras.layers.Dropout(dropout))
    model.add(keras.layers.Dense(1, activation='linear'))

    model.compile(optimizer=keras.optimizers.Adam(learning_rate=lr),
                  loss='mse',
                  metrics=['mae'])
    return model


def tune_lstm_manual(X_train, y_train, input_shape=None, epochs=15):
    """
    Manual hyperparameter tuning loop instead of keras-tuner.
    """
    param_grid = [
        {"units": 32, "dropout": 0.1, "lr": 0.001},
        {"units": 64, "dropout": 0.2, "lr": 0.001},
        {"units": 128, "dropout": 0.3, "lr": 0.0005},
    ]

    best_model = None
    best_mae = 1e12

    for params in param_grid:
        print(f"\nTrying LSTM params: {params}")

        model = build_lstm(
            units=params["units"],
            dropout=params["dropout"],
            lr=params["lr"],
            input_shape=input_shape
        )

        history = model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=32,
            validation_split=0.1,
            verbose=0
        )

        val_mae = history.history['val_mae'][-1]
        print(f"Validation MAE = {val_mae}")

        if val_mae < best_mae:
            best_mae = val_mae
            best_model = model

    print("\nBest LSTM found with MAE =", best_mae)
    return best_model


# -----------------------
# 2) Random Forest Grid Search
# -----------------------
def tune_random_forest(X_train, y_train):
    params = {
        "n_estimators": [100, 200],
        "max_depth": [10, 20, None],
        "min_samples_split": [2, 5]
    }

    rf = RandomForestRegressor(random_state=42)
    grid = GridSearchCV(rf, params, cv=3, scoring='neg_mean_absolute_error')
    grid.fit(X_train, y_train)
    return grid.best_estimator_


# -----------------------
# 3) XGBoost Grid Search
# -----------------------
def tune_xgboost(X_train, y_train):
    params = {
        "max_depth": [3, 5],
        "learning_rate": [0.05, 0.1],
        "n_estimators": [100, 200]
    }

    model = xgb.XGBRegressor(objective="reg:squarederror")
    grid = GridSearchCV(model, params, cv=3, scoring='neg_mean_absolute_error')
    grid.fit(X_train, y_train)
    return grid.best_estimator_


# -----------------------
# 4) Evaluation
# -----------------------
def evaluate_model(model, X, y):
    preds = model.predict(X)
    preds = preds.reshape(-1)
    rmse = np.sqrt(mean_squared_error(y, preds))
    mae = mean_absolute_error(y, preds)
    return rmse, mae, preds


# -----------------------
# 5) Visualization
# -----------------------
def plot_actual_vs_pred(y_true, preds, title="Actual vs Pred"):
    df = pd.DataFrame({
        "Actual": y_true,
        "Predicted": preds
    })
    fig = px.line(df, title=title)
    fig.show()


# -----------------------
# 6) Save models
# -----------------------
def save_model(model, name):
    os.makedirs("saved_models", exist_ok=True)

    if hasattr(model, "save"):
        path = f"saved_models/{name}.h5"
        model.save(path)
    else:
        path = f"saved_models/{name}.pkl"
        joblib.dump(model, path)

    print("Saved:", path)
    return path


# -----------------------
# 7) Main Pipeline
# -----------------------
def run_pipeline(X_train, y_train, X_test, y_test, X_holdout=None, y_holdout=None):
    input_shape = X_train.shape[1:]

    print("\n========== LSTM Tuning ==========")
    lstm_model = tune_lstm_manual(X_train, y_train, input_shape=input_shape)

    print("\n========== Random Forest Tuning ==========")
    rf_model = tune_random_forest(X_train, y_train)

    print("\n========== XGBoost Tuning ==========")
    xgb_model = tune_xgboost(X_train, y_train)

    print("\n========== Evaluation ==========")
    models = {"LSTM": lstm_model, "RF": rf_model, "XGB": xgb_model}
    scores = {}

    for name, model in models.items():
        rmse, mae, preds = evaluate_model(model, X_test, y_test)
        scores[name] = {"RMSE": rmse, "MAE": mae}
        print(f"{name}: RMSE={rmse:.3f}, MAE={mae:.3f}")

    # select best model
    best = min(scores, key=lambda k: scores[k]["RMSE"])
    print("\nBest model =", best)

    # save models
    for name, model in models.items():
        save_model(model, name)

    return models, scores
```

```python
import os
import numpy as np
import pandas as pd
import joblib
import tensorflow as tf # CRITICAL: Fix for TensorFlow logging
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import xgboost as xgb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# =======================================================
# 0. Data Loading and Preparation
# =======================================================
FILE_NAME = "/content/drive/MyDrive/FIFA_LSTM_Project/final_features_for_modeling.csv" # Using your uploaded file name

try:
    # --- Load Data ---
    df = pd.read_csv(FILE_NAME)

    # --- Feature and Target Selection ---
    # Dropping player identifiers and the target
    X = df.drop(columns=['full_name', 'name', 'birth_date', 'nationality', 'body_type',
                         'national_team', 'national_team_position', 'name_key',
                         'player_name_key', 'player_id_key', 'value_euro'],
                         errors='ignore').select_dtypes(include=np.number)
    y = df['value_euro']

    # --- Clean Data (Handle any residual NaNs) ---
    X = X.fillna(X.mean())
    y = y.fillna(y.mean())

    # --- Split Data: Training/Validation/Holdout Test ---
    X_train_val, X_holdout, y_train_val, y_holdout = train_test_split(X, y, test_size=0.1, random_state=42)
    X_train, X_test, y_train, y_test = train_test_split(X_train_val, y_train_val, test_size=0.2, random_state=42)

    # --- Scaling (Critical for LSTM) ---
    scaler_X = MinMaxScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)

    # --- Reshape for LSTM: (samples, 1, features) ---
    n_features = X_train_scaled.shape[1]
    X_train_lstm = X_train_scaled.reshape(X_train_scaled.shape[0], 1, n_features)
    X_test_lstm = X_test_scaled.reshape(X_test_scaled.shape[0], 1, n_features)

    print(f"Data ready for tuning. Total features: {n_features}")

except FileNotFoundError:
    print(f"FATAL ERROR: '{FILE_NAME}' not found. Check file name and path.")
    exit()


# =======================================================
# MILESTONE 6: Hyperparameter Tuning Functions
# =======================================================

def evaluate_model(model, X, y_true, name, X_test_lstm):
    """Evaluates the model on the provided data and returns metrics."""
    # Select the appropriate input shape for evaluation
    if name == 'LSTM':
        X_eval = X_test_lstm
    else:
        X_eval = X

    preds = model.predict(X_eval)
    preds = np.array(preds).reshape(-1)

    rmse = np.sqrt(mean_squared_error(y_true, preds))
    mae = mean_absolute_error(y_true, preds)
    r2 = r2_score(y_true, preds)

    return {"RMSE": rmse, "MAE": mae, "R2": r2, "Predictions": preds}

def tune_lstm_manual(X_train, y_train, input_shape, epochs=15):
    """Manual hyperparameter tuning loop for LSTM models."""
    print("\n--- Tuning LSTM Model ---")
    param_grid = [
        {"units": 32, "dropout": 0.1, "lr": 0.001},
        {"units": 64, "dropout": 0.2, "lr": 0.001},
        {"units": 128, "dropout": 0.3, "lr": 0.0005},
    ]

    best_model = None
    best_mae = np.inf

    for params in param_grid:
        tf.get_logger().setLevel('ERROR')

        model = Sequential([
            LSTM(params["units"], activation='tanh', input_shape=input_shape),
            Dropout(params["dropout"]),
            Dense(1, activation='linear')
        ])
        model.compile(optimizer=Adam(learning_rate=params["lr"]), loss='mse', metrics=['mae'])

        history = model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=32,
            validation_split=0.1,
            verbose=0
        )

        val_mae = history.history['val_mae'][-1]
        print(f"LSTM Config (Units:{params['units']}, LR:{params['lr']}): Val MAE = {val_mae:.4f}")

        if val_mae < best_mae:
            best_mae = val_mae
            best_model = model

    tf.get_logger().setLevel('WARNING')
    print(f" Best LSTM selected based on Validation MAE: {best_mae:.4f}")
    return best_model

def tune_random_forest(X_train, y_train):
    """GridSearchCV for Random Forest."""
    print("\n--- Tuning Random Forest (RF) Model ---")
    params = {
        "n_estimators": [100, 200],
        "max_depth": [10, 20],
        "min_samples_split": [2, 5]
    }
    rf = RandomForestRegressor(random_state=42, n_jobs=-1)
    grid = GridSearchCV(rf, params, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
    grid.fit(X_train, y_train)
    print(f" RF Best Params: {grid.best_params_}")
    return grid.best_estimator_

def tune_xgboost(X_train, y_train):
    """GridSearchCV for XGBoost."""
    print("\n--- Tuning XGBoost Model ---")
    params = {
        "max_depth": [3, 5],
        "learning_rate": [0.05, 0.1],
        "n_estimators": [100, 200]
    }
    model = xgb.XGBRegressor(objective="reg:squarederror", random_state=42, n_jobs=-1)
    grid = GridSearchCV(model, params, cv=3, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)
    grid.fit(X_train, y_train)
    print(f" XGB Best Params: {grid.best_params_}")
    return grid.best_estimator_


# =======================================================
# MILESTONE 6: Execution, Selection & Validation
# =======================================================

print("--------------------------------------------------")
print("EXECUTING MILESTONE 6: Tuning & Validation 🛠️")
print("--------------------------------------------------")

# --- Run Tuning ---
tuned_lstm = tune_lstm_manual(X_train_lstm, y_train, input_shape=(1, n_features))
tuned_rf = tune_random_forest(X_train, y_train)
tuned_xgb = tune_xgboost(X_train, y_train)

models = {"LSTM": tuned_lstm, "RandomForest": tuned_rf, "XGBoost": tuned_xgb}
eval_results = pd.DataFrame(columns=['RMSE', 'MAE', 'R2'])

# --- Validation Results (Testing on holdout datasets: X_test) ---
print("\n========== FINAL VALIDATION RESULTS (on Test Set) ==========")
for name, model in models.items():
    # Pass X_test_lstm specifically for LSTM evaluation
    results = evaluate_model(model, X_test, y_test, name, X_test_lstm)
    eval_results.loc[name] = [results['RMSE'], results['MAE'], results['R2']]
    print(f"Model: {name}")
    print(f"  > RMSE: €{results['RMSE']:.2f}")
    print(f"  > MAE: €{results['MAE']:.2f}")
    print(f"  > R2: {results['R2']:.4f}\n")

# --- Final Model Selection ---
best_model_name = eval_results['RMSE'].idxmin()
best_model = models[best_model_name]

print("--------------------------------------------------")
print(f" FINAL MODEL SELECTED (Lowest RMSE): **{best_model_name}**")
print("--------------------------------------------------")

# --- Save Best Model and Scaler ---
save_dir = "saved_models"
os.makedirs(save_dir, exist_ok=True)

if best_model_name == 'LSTM':
    best_model.save(os.path.join(save_dir, f"{best_model_name}_final.h5"))
else:
    joblib.dump(best_model, os.path.join(save_dir, f"{best_model_name}_final.pkl"))

joblib.dump(scaler_X, os.path.join(save_dir, 'scaler_X.pkl'))

print(f" Selected model and scaler saved to '{save_dir}/' for Week 8 deployment.")
```
