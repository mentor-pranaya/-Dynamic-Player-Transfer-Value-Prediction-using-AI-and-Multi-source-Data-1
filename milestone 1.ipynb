```python
import os
import requests
os.makedirs("statsbomb_data", exist_ok=True)
url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/competitions.json"
r = requests.get(url)
with open("statsbomb_data/competitions.json", "wb") as f:
    f.write(r.content)

print("Downloaded competitions.json")
```

```python
import pandas as pd
import json

with open("statsbomb_data/competitions.json") as f:
    competitions = json.load(f)

df_competitions = pd.DataFrame(competitions)
print(df_competitions.head())
print(df_competitions.info())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

url = "https://www.transfermarkt.com/spieler-statistik/wertvollstespieler"
headers = {"User-Agent": "Mozilla/5.0"}
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, "html.parser")

players = []
for row in soup.select('.items tr')[1:6]:  # Sample: first 5 players
    cols = row.find_all('td')
    if len(cols) > 4:
        name = cols[1].get_text(strip=True)
        market_value = cols[-1].get_text(strip=True)
        players.append({"name": name, "market_value": market_value})

df_transfermarkt = pd.DataFrame(players)
print(df_transfermarkt)
```

```python
import pandas as pd
from textblob import TextBlob

tweets = [
    {"text": "Player X was amazing today!"},
    {"text": "Player Y was disappointing."},
    {"text": "Player Z is injured again."}
]
df_tweets = pd.DataFrame(tweets)
df_tweets['polarity'] = df_tweets['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
print(df_tweets)
```
