# 🔧 03 — Feature Engineering
Create performance-derived features, merge sentiment and injury features.

```python
import pandas as pd
from pathlib import Path
DATA = Path('..') / 'data' / 'processed'
FEATURE_OUT = Path('..') / 'data' / 'interim' / 'feature_set.csv'
FEATURE_OUT.parent.mkdir(parents=True, exist_ok=True)

perf = pd.read_csv(DATA / 'clean_statsbomb.csv')
sent = pd.read_csv(DATA / 'clean_sentiment.csv')
inj = pd.read_csv(DATA / 'clean_injuries.csv')
print('Loaded processed tables')
```

## Create per-event normalized features

```python
df = perf.copy()
for c in ['shots','passes','tackles','events']:
    if c not in df.columns:
        df[c] = 0
df['shots_per_event'] = df['shots'] / (df['events'] + 1)
df['passes_per_event'] = df['passes'] / (df['events'] + 1)
df['tackles_per_event'] = df['tackles'] / (df['events'] + 1)
df['activity_index'] = df['events'] / (df['events'].max() + 1)
display(df.head())
```

## Merge sentiment and injuries

```python
df = df.merge(sent, on='player_name', how='left')
df = df.merge(inj, on='player_name', how='left')
df = df.fillna({'sentiment_score':0, 'days_absent':0})
df.to_csv(FEATURE_OUT, index=False)
print('Saved features to', FEATURE_OUT)
display(df.head())
```
