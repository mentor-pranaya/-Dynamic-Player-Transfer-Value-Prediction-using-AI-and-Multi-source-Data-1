# 📊 01 — Exploratory Data Analysis (EDA)
Explore sample TransferIQ datasets in `data/raw/`.

Files used:
- `data/raw/performance.csv`
- `data/raw/transfermarkt_values.csv`
- `data/raw/tweets.csv`
- `data/raw/injuries.csv`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style='whitegrid')
ROOT = os.path.join('..','data','raw')
perf = pd.read_csv(os.path.join(ROOT, 'performance.csv'))
values = pd.read_csv(os.path.join(ROOT, 'transfermarkt_values.csv'))
tweets = pd.read_csv(os.path.join(ROOT, 'tweets.csv'))
inj = pd.read_csv(os.path.join(ROOT, 'injuries.csv'))

print('Loaded tables:')
print('performance', perf.shape)
print('values', values.shape)
print('tweets', tweets.shape)
print('injuries', inj.shape)
```

## Basic previews

```python
display(perf.head())
display(values.head())
display(tweets.head())
display(inj.head())
```

## Missing values and types

```python
for name, df in [('perf', perf), ('values', values), ('tweets', tweets), ('inj', inj)]:
    print(f"--- {name} ---")
    print(df.dtypes)
    print(df.isna().sum())
    print()
```

## Market value distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(values['market_value_eur'].dropna(), bins=20)
plt.title('Market value (EUR) distribution')
plt.show()
```

## Market value by player (bar)

```python
plt.figure(figsize=(8,4))
sns.barplot(data=values, x='player_name', y='market_value_eur')
plt.xticks(rotation=40)
plt.title('Market value by player')
plt.show()
```

## Sentiment overview

```python
sent = tweets.groupby('player_keyword')['vader_compound'].mean().reset_index()
plt.figure(figsize=(8,3))
sns.barplot(data=sent, x='player_keyword', y='vader_compound')
plt.xticks(rotation=40)
plt.title('Average tweet sentiment (VADER) per player')
plt.show()
```
