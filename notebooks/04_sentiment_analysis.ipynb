# 💬 04 — Sentiment Analysis (VADER)
Demonstrate sentiment scoring and aggregation over tweets.
If you used `src/data_collection/twitter_sentiment.py` earlier, the raw file is `data/raw/tweets.csv`.

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from pathlib import Path
RAW = Path('..') / 'data' / 'raw'
tweets = pd.read_csv(RAW / 'tweets.csv')
analyzer = SentimentIntensityAnalyzer()
print('Loaded tweets:', tweets.shape)
```

## Score tweets (if not already present)

```python
if 'vader_compound' not in tweets.columns:
    tweets['vader_compound'] = tweets['tweet_text'].astype(str).apply(lambda t: analyzer.polarity_scores(t)['compound'])
tweets['tweet_date'] = pd.to_datetime(tweets['tweet_date'], errors='coerce')
display(tweets.head())
```

## Aggregate sentiment by player and save to processed

```python
sent = tweets.groupby('player_keyword')['vader_compound'].mean().reset_index()
sent = sent.rename(columns={'player_keyword':'player_name','vader_compound':'sentiment_score'})
proc = Path('..') / 'data' / 'processed'
proc.mkdir(parents=True, exist_ok=True)
sent.to_csv(proc / 'clean_sentiment.csv', index=False)
print('Saved aggregated sentiment to', proc / 'clean_sentiment.csv')
display(sent)
```
