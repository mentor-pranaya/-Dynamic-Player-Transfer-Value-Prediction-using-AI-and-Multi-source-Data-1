```python
import pandas as pd
import numpy as np

# --- Collect player performance data from StatsBomb Open Data (download example) ---
import os
import requests
os.makedirs('data', exist_ok=True)
url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/competitions.json"
r = requests.get(url)
with open('data/competitions.json', 'wb') as f:
    f.write(r.content)

with open('data/competitions.json') as f:
    competitions = pd.read_json(f)
print("StatsBomb performance sample:\n", competitions.head())

# --- Scrape Market Value Data from Transfermarkt (demo data shown, scraping for illustration) ---
market_data = pd.DataFrame({
    'player': ['Player X', 'Player Y', 'Player Z'],
    'market_value': [6.2, 7.8, 4.5],
    'position': ['Midfielder', 'Forward', 'Defender']
})

print("\nTransfermarkt market value sample:\n", market_data)

# --- Fetch Twitter data and perform sentiment analysis ---
tweets = pd.DataFrame({
    'text': [
        "Player X was fantastic today!",
        "Sad about Player Y's injury.",
        "Player Z's transfer is rumored."
    ]
})
from textblob import TextBlob
tweets['polarity'] = tweets['text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
print("\nSample tweet sentiment:\n", tweets)

# --- Injury history (sample) ---
injury_data = pd.DataFrame({
    'player': ['Player X', 'Player Y', 'Player Z'],
    'injury_type': ['Hamstring', 'Ankle', 'ACL'],
    'daysmissed': [20, 34, 120]
})
print("\nInjury history sample:\n", injury_data)

# --- Initial Data Exploration ---
print("\nMarket Value Describe:\n", market_data.describe())
print("\nMissing Data in StatsBomb:\n", competitions.isnull().sum())
print("\nInjury Data Distributions:\n", injury_data.describe())

# --- Exploration Report ---
report = {
    'statsbomb_shape': competitions.shape,
    'market_shape': market_data.shape,
    'injury_shape': injury_data.shape,
    'market_missing': market_data.isnull().sum().to_dict(),
    'injury_missing': injury_data.isnull().sum().to_dict(),
    'statsbomb_missing': competitions.isnull().sum().to_dict()
}
print("\nExploration report:", report)
```
