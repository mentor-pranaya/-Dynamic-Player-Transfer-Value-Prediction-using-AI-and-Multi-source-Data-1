**DATA CLEANING -WEEK 2**

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('/content/fifa_players.csv')

# Strip whitespace from column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Drop completely empty rows
df.dropna(how='all', inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Convert date columns
if 'birth_date' in df.columns:
    df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')

# Convert numeric columns
numeric_cols = [
    'age', 'height_cm', 'weight_kgs', 'overall_rating', 'potential',
    'value_euro', 'wage_euro', 'international_reputation(1-5)',
    'weak_foot(1-5)', 'skill_moves(1-5)', 'release_clause_euro',
    'national_rating', 'national_jersey_number', 'crossing', 'finishing',
    'heading_accuracy', 'short_passing', 'volleys', 'dribbling', 'curve',
    'freekick_accuracy', 'long_passing', 'ball_control', 'acceleration',
    'sprint_speed', 'agility', 'reactions', 'balance', 'shot_power',
    'jumping', 'stamina', 'strength', 'long_shots', 'aggression',
    'interceptions', 'positioning', 'vision', 'penalties', 'composure',
    'marking', 'standing_tackle', 'sliding_tackle'
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill missing categorical values
categorical_cols = [
    'name', 'full_name', 'positions', 'nationality', 'preferred_foot',
    'body_type', 'national_team', 'national_team_position'
]

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')

# Optional: Normalize currency columns (e.g., convert to millions)
for col in ['value_euro', 'wage_euro', 'release_clause_euro']:
    if col in df.columns:
        df[col] = df[col] / 1e6  # Convert to millions

#  now Saving cleaned data
df.to_csv('fifa_players_cleaned.csv', index=False)
```

DATA CLEANING FOR GAME EVENTS DATASETS
{dataset has some non-integer numeric values (like NaN or decimals) in the supposed integer ID columns (for example, "player_id", "club_id", etc.).
When we try to convert them to Int64, pandas complains because it can’t safely cast float → int if there are invalid entries.}

```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
df = pd.read_csv('/content/game_events.csv')

# Step 2: Basic Overview
print("Before cleaning:\n")
print(df.info())
print("\nFirst 5 rows:\n", df.head())

# Step 3: Handle Missing Values
# Fill missing text columns with 'Unknown'
text_cols = ['type', 'description']
for col in text_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')

# Step 4: Convert 'date' column to datetime
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Step 5: Convert ID columns safely
id_cols = ['game_event_id', 'game_id', 'club_id', 'player_id', 'player_in_id', 'player_assist_id']

for col in id_cols:
    if col in df.columns:
        # Convert to numeric, replace non-numeric with NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')
        # Fill NaN with -1, but keep as float to avoid casting issues
        df[col] = df[col].fillna(-1)

# Step 6: Handle minute column
if 'minute' in df.columns:
    df['minute'] = pd.to_numeric(df['minute'], errors='coerce').fillna(0)
    # Ensure realistic range (0–130)
    df.loc[(df['minute'] < 0) | (df['minute'] > 130), 'minute'] = np.nan

# Step 7: Remove duplicates
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"\nRemoved {before - after} duplicate rows.")

# Step 8: Check for remaining missing values
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Step 9: Dataset summary
print("\nSummary after cleaning:")
print(df.describe(include='all'))

# Step 10: Save cleaned file
df.to_csv('game_events_cleaned.csv', index=False)
print("\n Cleaned dataset saved as 'game_events_cleaned.csv'")
```

cleaning for injury dataset


```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
df = pd.read_csv('/content/dataset.csv')

# Step 2: Quick overview
print("Before Cleaning:\n")
print(df.info())
print("\nFirst 5 rows:\n", df.head())

# Step 3: Convert date column to datetime
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')

# Step 4: Handle missing values

# Fill missing categorical (object) columns
df['position'] = df['position'].fillna('Unknown')

# Fill numeric columns with appropriate strategies
num_cols_mean = ['pace', 'physic']  # few missing, can use mean
for col in num_cols_mean:
    df[col] = df[col].fillna(df[col].mean())

# Columns with large missing data: fill with 0 (represents unavailable previous data)
large_missing_cols = [
    'cumulative_minutes_played', 'cumulative_games_played', 'minutes_per_game_prev_seasons',
    'avg_days_injured_prev_seasons', 'avg_games_per_season_prev_seasons',
    'significant_injury_prev_season', 'cumulative_days_injured',
    'season_days_injured_prev_season'
]
df[large_missing_cols] = df[large_missing_cols].fillna(0)

# Fill missing position_numeric (use median for balance)
df['position_numeric'] = df['position_numeric'].fillna(df['position_numeric'].median())

# Step 5: Handle invalid or inconsistent values

# Ensure no negative numbers in gameplay/injury stats
cols_to_check = [
    'season_days_injured', 'total_days_injured', 'season_minutes_played',
    'season_games_played', 'season_matches_in_squad', 'total_minutes_played',
    'total_games_played', 'age', 'height_cm', 'weight_kg', 'fifa_rating', 'bmi'
]
for col in cols_to_check:
    df[col] = df[col].apply(lambda x: np.nan if x < 0 else x)

# Step 6: Drop duplicates if any
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"\nRemoved {before - after} duplicate rows.")

# Step 7: Recheck missing values
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Step 8: Clean text columns (trim spaces, standardize)
df['nationality'] = df['nationality'].str.strip()
df['work_rate'] = df['work_rate'].str.strip().str.title()
df['position'] = df['position'].str.strip().str.title()

# Step 9: Verify numeric data types
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Step 10: Final dataset summary
print("\nAfter Cleaning:\n")
print(df.describe(include='all'))

# Step 11: Save cleaned dataset
df.to_csv('fifa_players_injury_cleaned.csv', index=False)
print("\n Cleaned dataset saved as 'fifa_players_injury_cleaned.csv'")
```

```python
# --------------------------------
# Step 1: Import Libraries
# --------------------------------
import pandas as pd
import numpy as np
from textblob import TextBlob

# --------------------------------
# Step 2: Load CSVs (No Header Fix)
# --------------------------------
train_path = '/content/twitter_training.csv'
val_path = '/content/twitter_validation.csv'

# These files don’t have headers, so we’ll add our own column names
column_names = ['id', 'entity', 'sentiment', 'text']

train_df = pd.read_csv(train_path, names=column_names, encoding='utf-8')
val_df = pd.read_csv(val_path, names=column_names, encoding='utf-8')

# --------------------------------
# Step 3: Basic Info
# --------------------------------
print("Training Data Info:\n")
print(train_df.info())
print("\nSample Rows:\n")
print(train_df.head())

# --------------------------------
# Step 4: Clean Data
# --------------------------------
train_df.dropna(subset=['text'], inplace=True)
val_df.dropna(subset=['text'], inplace=True)
train_df.drop_duplicates(subset=['text'], inplace=True)
val_df.drop_duplicates(subset=['text'], inplace=True)

# --------------------------------
# Step 5: Sentiment Analysis
# --------------------------------
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

train_df['sentiment_score'] = train_df['text'].apply(get_sentiment)
val_df['sentiment_score'] = val_df['text'].apply(get_sentiment)

# --------------------------------
# Step 6: Label Sentiments
# --------------------------------
def label_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

train_df['sentiment_label'] = train_df['sentiment_score'].apply(label_sentiment)
val_df['sentiment_label'] = val_df['sentiment_score'].apply(label_sentiment)

# --------------------------------
# Step 7: Summary
# --------------------------------
print("\nSentiment Distribution (Training):\n", train_df['sentiment_label'].value_counts())
print("\nSentiment Distribution (Validation):\n", val_df['sentiment_label'].value_counts())

# --------------------------------
# Step 8: Preview Results
# --------------------------------
print("\nSample Output:\n")
print(train_df[['text', 'sentiment_score', 'sentiment_label']].head())
```

```python
"""
Advanced Feature Engineering and Sentiment Analysis
---------------------------------------------------
This script:
 - Asks for dataset file paths interactively.
 - Loads player, appearance, injury, transfer, and sentiment data.
 - Creates advanced features:
     * Rolling averages & trends for performance.
     * Injury impact on market value.
     * Sentiment scores from social media data.
 - Saves final datasets:
     * performance_features.csv
     * injury_impact_features.csv
     * sentiment_aggregates.csv
     * final_feature_set.csv
     * sentiment_report.md
"""

import os
import pandas as pd
import numpy as np
from scipy.stats import linregress
from datetime import datetime

# --------------------- Step 1: Ask for file paths -------------------------
print("📂 Please enter full paths for your data files (press Enter to skip any missing file):\n")

def ask_path(name):
    path = input(f"→ Path for {name} dataset: ").strip()
    if path and os.path.exists(path):
        print(f"  ✅ Found {name} file at {path}")
        return path
    elif path:
        print(f"  ⚠️  File not found at {path}, skipping...")
    else:
        print(f"  ⏭️  Skipped {name}")
    return None

paths = {
    'appearances': ask_path('Appearances'),
    'players': ask_path('Players'),
    'injuries': ask_path('Injuries'),
    'transfers': ask_path('Transfers'),
    'game_events': ask_path('Game Events'),
    'sentiment_raw': ask_path('Sentiment (Social Media)')
}

# --------------------- Step 2: Read datasets safely -----------------------
def try_read_csv(path):
    try:
        return pd.read_csv(path)
    except Exception as e:
        print(f"❌ Could not read {path}: {e}")
        return None

df_appear = try_read_csv(paths['appearances']) if paths['appearances'] else None
df_players = try_read_csv(paths['players']) if paths['players'] else None
df_injuries = try_read_csv(paths['injuries']) if paths['injuries'] else None
df_transfers = try_read_csv(paths['transfers']) if paths['transfers'] else None
df_sent_raw = try_read_csv(paths['sentiment_raw']) if paths['sentiment_raw'] else None

print("\n✅ All available datasets have been loaded successfully (if paths were correct).")

# --------------------- Step 3: Sentiment Utilities -----------------------
try:
    from nltk.sentiment import SentimentIntensityAnalyzer
    import nltk
    nltk.download('vader_lexicon', quiet=True)
    _nltk_ok = True
except Exception:
    _nltk_ok = False

try:
    from textblob import TextBlob
    _textblob_ok = True
except Exception:
    _textblob_ok = False

# fallback simple lexicon
SIMPLE_LEXICON = {
    'good':2, 'great':3, 'excellent':4, 'amazing':4, 'love':3, 'like':1,
    'bad':-2, 'terrible':-3, 'awful':-3, 'hate':-3, 'poor':-2, 'injury':-1
}

def simple_sentiment(text):
    if not isinstance(text, str):
        return 0.0
    score = 0
    for word in text.lower().split():
        score += SIMPLE_LEXICON.get(word.strip(".,!?;\"'"), 0)
    return float(score)

def compute_sentiment_scores(series):
    df = pd.DataFrame({'text': series.astype(str)})
    if _nltk_ok:
        sia = SentimentIntensityAnalyzer()
        df['sent_compound'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
        df['sent_pos'] = df['text'].apply(lambda x: sia.polarity_scores(x)['pos'])
        df['sent_neg'] = df['text'].apply(lambda x: sia.polarity_scores(x)['neg'])
        df['method'] = 'vader'
        return df
    elif _textblob_ok:
        df['sent_polarity'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
        df['sent_subjectivity'] = df['text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
        df['method'] = 'textblob'
        return df
    else:
        df['sent_simple'] = df['text'].apply(simple_sentiment)
        df['method'] = 'simple'
        return df

# --------------------- Step 4: Feature Engineering -----------------------
def rolling_performance_features(df):
    df = df.copy()
    if 'player_id' not in df.columns and 'p_id2' in df.columns:
        df.rename(columns={'p_id2':'player_id'}, inplace=True)

    if 'date' not in df.columns:
        for c in ['match_date', 'game_date', 'datetime']:
            if c in df.columns:
                df.rename(columns={c:'date'}, inplace=True)
                break
    df['date'] = pd.to_datetime(df['date'], errors='coerce')

    metrics = [c for c in ['minutes','goals','assists','rating'] if c in df.columns]
    df = df.sort_values(['player_id','date'])
    result = []
    for pid, grp in df.groupby('player_id'):
        g = grp.reset_index(drop=True)
        for w in [3,5]:
            for m in metrics:
                g[f'{m}_roll_mean_{w}'] = g[m].rolling(w, min_periods=1).mean()
        for m in metrics:
            if g[m].notna().sum() >= 2:
                x = np.arange(len(g))
                slope = linregress(x, g[m].fillna(0)).slope
            else:
                slope = 0
            g[f'{m}_trend_slope'] = slope
        result.append(g)
    return pd.concat(result, ignore_index=True)

def injury_impact_features(df_players, df_injuries):
    if df_injuries is None or df_injuries.empty:
        return pd.DataFrame()
    df_players['date'] = pd.to_datetime(df_players.get('date'), errors='coerce')
    results = []
    for _, row in df_injuries.iterrows():
        pid = row.get('player_id') or row.get('player_name') or row.get('p_id2')
        inj_date = pd.to_datetime(row.get('injury_date') or row.get('date'), errors='coerce')
        if pd.isna(pid) or pd.isna(inj_date): continue
        sub = df_players[df_players['player_id']==pid]
        pre = sub[(sub['date']<inj_date) & (sub['date']>=inj_date - pd.Timedelta(days=180))]
        post = sub[(sub['date']>inj_date) & (sub['date']<=inj_date + pd.Timedelta(days=180))]
        pre_val, post_val = pre['market_value'].mean(), post['market_value'].mean()
        delta = post_val - pre_val if not np.isnan(pre_val) and not np.isnan(post_val) else np.nan
        pct = (delta/pre_val*100) if pre_val and not np.isnan(pre_val) else np.nan
        results.append({'player_id':pid,'injury_date':inj_date,'pre_value':pre_val,
                        'post_value':post_val,'delta':delta,'pct_change':pct})
    return pd.DataFrame(results)

# --------------------- Step 5: Run Analysis ------------------------------
outputs = {}

# Rolling performance
if df_appear is not None:
    perf = rolling_performance_features(df_appear)
    perf.to_csv('performance_features.csv', index=False)
    outputs['perf'] = perf
    print("✅ Saved rolling performance features → performance_features.csv")

# Injury impact
if df_players is not None and df_injuries is not None:
    inj = injury_impact_features(df_players, df_injuries)
    inj.to_csv('injury_impact_features.csv', index=False)
    outputs['inj'] = inj
    print("✅ Saved injury impact features → injury_impact_features.csv")

# Sentiment analysis
if df_sent_raw is not None and 'text' in df_sent_raw.columns:
    sent_scores = compute_sentiment_scores(df_sent_raw['text'])
    merged = df_sent_raw.join(sent_scores)
    if 'player_id' in merged.columns:
        agg = merged.groupby('player_id').agg('mean').reset_index()
        agg.to_csv('sentiment_aggregates.csv', index=False)
        outputs['sent'] = agg
        print("✅ Saved sentiment aggregates → sentiment_aggregates.csv")
    else:
        merged.to_csv('sentiment_per_text.csv', index=False)
        outputs['sent'] = merged
        print("✅ Saved per-text sentiment → sentiment_per_text.csv")
else:
    print("⚠️ No sentiment text data found or missing 'text' column.")

# --------------------- Step 6: Combine Final Feature Set -----------------
final = None
if 'perf' in outputs:
    latest = outputs['perf'].sort_values(['player_id','date']).groupby('player_id').tail(1)
    final = latest.copy()
elif df_players is not None:
    final = df_players.copy()

if final is not None:
    if 'inj' in outputs:
        final = final.merge(outputs['inj'], on='player_id', how='left')
    if 'sent' in outputs and isinstance(outputs['sent'], pd.DataFrame):
        if 'player_id' in outputs['sent'].columns:
            final = final.merge(outputs['sent'], on='player_id', how='left')
    final.to_csv('final_feature_set.csv', index=False)
    print("✅ Final feature set saved → final_feature_set.csv")

# --------------------- Step 7: Create Sentiment Report -------------------
report = []
report.append("# Sentiment Analysis Report")
report.append(f"Generated: {datetime.utcnow().isoformat()}Z\n")

if 'sent' in outputs:
    report.append("## Overview of sentiment data")
    report.append(str(outputs['sent'].describe()))
else:
    report.append("No sentiment data available for report.")

with open('sentiment_report.md','w',encoding='utf-8') as f:
    f.write("\n\n".join(report))
print("📄 Sentiment report saved → sentiment_report.md")

print("\n🎯 Task complete: All available features and reports have been generated successfully.")
```

```python
# ===========================================
# 📘 Advanced Feature Engineering + Sentiment Analysis
# ===========================================

# ✅ 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from textblob import TextBlob

# Optional: use any available Matplotlib style (no seaborn dependency)
plt.style.use('ggplot')

# ===========================================
# ✅ 2. Load Datasets
# ===========================================

# Replace these file names with your actual file paths
fifa_df = pd.read_csv("fifa_players_cleaned.csv")
injury_df = pd.read_csv("injury_data.csv")
appearance_df = pd.read_csv("appearances.csv")
twitter_df = pd.read_csv("twitter_data.csv")   # if you have one

print("✅ Data loaded successfully!\n")
print("FIFA Shape:", fifa_df.shape)
print("Injury Shape:", injury_df.shape)
print("Appearance Shape:", appearance_df.shape)

# ===========================================
# ✅ 3. Feature Engineering (Performance Trends)
# ===========================================

# Example: Performance trend = moving average of overall_rating
fifa_df['rating_trend'] = fifa_df['overall_rating'].rolling(window=5, min_periods=1).mean()

# Injury impact on player value
injury_impact = injury_df.groupby('p_id2')['total_days_injured'].mean().reset_index()
injury_impact.columns = ['player_id', 'avg_days_injured']

# Merge with FIFA dataset on player name or ID (adjust accordingly)
fifa_df = fifa_df.merge(injury_impact, left_on='full_name', right_on='player_id', how='left')

# Fill missing injury data with 0
fifa_df['avg_days_injured'] = fifa_df['avg_days_injured'].fillna(0)

# Create new feature: Value drop due to injury
fifa_df['injury_value_impact'] = fifa_df['avg_days_injured'] * 0.001 * fifa_df['value_euro']

# ===========================================
# ✅ 4. Sentiment Analysis (Twitter / Social Media Data)
# ===========================================

def get_sentiment(text):
    try:
        return TextBlob(text).sentiment.polarity
    except:
        return 0

# Calculate sentiment polarity
twitter_df['sentiment_score'] = twitter_df['Text'].apply(get_sentiment)

# Aggregate sentiment by player
player_sentiment = twitter_df.groupby('Player')['sentiment_score'].mean().reset_index()
player_sentiment.columns = ['full_name', 'avg_sentiment']

# Merge with FIFA dataset
fifa_df = fifa_df.merge(player_sentiment, on='full_name', how='left')
fifa_df['avg_sentiment'] = fifa_df['avg_sentiment'].fillna(0)

# ===========================================
# ✅ 5. Generate Final Feature Set for Modeling
# ===========================================

final_features = fifa_df[[
    'full_name', 'age', 'height_cm', 'weight_kgs',
    'overall_rating', 'potential', 'value_euro',
    'wage_euro', 'rating_trend', 'avg_days_injured',
    'injury_value_impact', 'avg_sentiment'
]]

print("\n✅ Final Feature Set:")
print(final_features.head())

# ===========================================
# ✅ 6. Visual Insights
# ===========================================

# Sentiment vs Market Value
plt.figure(figsize=(8,5))
plt.scatter(final_features['avg_sentiment'], final_features['value_euro'], alpha=0.5)
plt.title("Sentiment Score vs Player Market Value")
plt.xlabel("Average Sentiment")
plt.ylabel("Market Value (Euros)")
plt.show()

# Injury vs Value
plt.figure(figsize=(8,5))
plt.scatter(final_features['avg_days_injured'], final_features['value_euro'], alpha=0.5, color='orange')
plt.title("Injury Impact on Player Value")
plt.xlabel("Average Days Injured")
plt.ylabel("Market Value (Euros)")
plt.show()

# ===========================================
# ✅ 7. Save Final Feature Dataset
# ===========================================

final_features.to_csv("final_features_for_modeling.csv", index=False)
print("\n💾 Final dataset saved as 'final_features_for_modeling.csv'")

# ===========================================
# ✅ 8. Report Insights
# ===========================================

print("\n📊 Sentiment Analysis Report:")
print("Average Sentiment Score across all players:", round(twitter_df['sentiment_score'].mean(), 3))
print("Players with positive sentiment:", (twitter_df['sentiment_score'] > 0).sum())
print("Players with negative sentiment:", (twitter_df['sentiment_score'] < 0).sum())

print("\n✅ Advanced Feature Engineering and Sentiment Analysis Completed Successfully!")
```

```python
# ======================
# Advanced Feature Engineering + Sentiment Analysis
# Complete notebook code block — run in Colab / Jupyter
# ======================

# STEP A: Install required packages (Colab only)
# (If running locally, install these with pip if not already installed)
!pip install -q vaderSentiment textblob rapidfuzz wordcloud

# Download textblob corpora (if needed)
!python -m textblob.download_corpora >/dev/null 2>&1

# ----------------------
# STEP 1: Imports
# ----------------------
import pandas as pd
import numpy as np
import re
from datetime import datetime
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from rapidfuzz import process, fuzz
from scipy.stats import linregress
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
sns.set_style("whitegrid")

# ----------------------
# STEP 2: File paths (update if needed)
# ----------------------
# Use the cleaned filenames you mentioned earlier
FIFA_FILE = '/content/fifa_players_cleaned.csv'
INJURY_FILE = '/content/fifa_players_injury_cleaned.csv'
APPEAR_FILE = '/content/appearance_data_cleaned.csv'     # appearance dataset
VALUATION_FILE = '/content/player_valuations.csv'     # valuation timeseries
TW_TRAIN = '/content/twitter_training.csv'
TW_VAL = '/content/twitter_validation.csv'

# ----------------------
# STEP 3: Load datasets safely
# ----------------------
def safe_read(path, **kwargs):
    try:
        return pd.read_csv(path, **kwargs)
    except Exception as e:
        print(f"Warning: could not read {path}: {e}")
        return pd.DataFrame()

fifa = safe_read(FIFA_FILE, encoding='utf-8')
injury = safe_read(INJURY_FILE, encoding='utf-8')
appearance = safe_read(APPEAR_FILE, encoding='utf-8')
valuation = safe_read(VALUATION_FILE, encoding='utf-8')
tw_train = safe_read(TW_TRAIN, header=None, encoding='utf-8')  # headerless
tw_val = safe_read(TW_VAL, header=None, encoding='utf-8')

print("Shapes loaded — fifa:", fifa.shape, "injury:", injury.shape, "appearance:", appearance.shape, "valuation:", valuation.shape,
      "tw_train:", tw_train.shape, "tw_val:", tw_val.shape)

# ----------------------
# STEP 4: Prepare Twitter combined dataset (headerless -> assign columns)
# ----------------------
# Many of your twitter files were headerless and had 4 columns: ID, Entity, Sentiment, Text
def prepare_twitter(df):
    if df.empty:
        return df
    # if it already has column names like 'Text' or 'text', keep them
    cols = df.columns.tolist()
    # Typical case: headerless with 4 columns
    if len(cols) >= 4 and all(isinstance(c, (int, np.integer)) for c in cols[:4]):
        df = df.iloc[:, :4]  # take first 4 columns
        df.columns = ['ID', 'Entity', 'Label', 'Text']
    else:
        # try to standardize by finding a column likely to be text
        text_col = None
        for c in df.columns:
            if str(c).lower() in ['text', 'tweet', 'body']:
                text_col = c; break
        if text_col is None:
            # fallback: treat last object column as text
            obj_cols = df.select_dtypes(include='object').columns.tolist()
            if obj_cols:
                text_col = obj_cols[-1]
        # ensure at least Text column exists
        if text_col:
            df = df.rename(columns={text_col: 'Text'})
        # create entity if missing
        if 'Entity' not in df.columns:
            # try second column if exists
            if df.shape[1] >= 2:
                df = df.rename(columns={df.columns[1]: 'Entity'})
            else:
                df['Entity'] = 'unknown'
    return df

tw_train = prepare_twitter(tw_train)
tw_val = prepare_twitter(tw_val)
twitter = pd.concat([tw_train, tw_val], ignore_index=True)
twitter = twitter.loc[:, twitter.columns.drop_duplicates()]  # drop accidental dup cols
print("Twitter combined shape:", twitter.shape)
print("Twitter columns:", twitter.columns.tolist())
display(twitter.head(3))

# ----------------------
# STEP 5: Clean Twitter text and compute VADER + TextBlob scores
# ----------------------
def clean_text(s):
    s = str(s)
    s = s.replace('\n', ' ')
    s = re.sub(r'http\S+|www\.\S+', '', s)
    s = re.sub(r'@\w+', '', s)
    s = re.sub(r'#[A-Za-z0-9_]+', '', s)
    s = re.sub(r'[^A-Za-z0-9\s]', ' ', s)
    s = re.sub(r'\s+', ' ', s).strip().lower()
    return s

if 'Text' not in twitter.columns:
    twitter['Text'] = twitter.iloc[:, -1].astype(str)

twitter['clean_text'] = twitter['Text'].apply(clean_text)

# VADER + TextBlob
vader = SentimentIntensityAnalyzer()
twitter['vader_score'] = twitter['clean_text'].apply(lambda t: vader.polarity_scores(t)['compound'])
twitter['textblob_polarity'] = twitter['clean_text'].apply(lambda t: TextBlob(t).sentiment.polarity)
# combined robust score (mean)
twitter['sentiment_score'] = twitter[['vader_score','textblob_polarity']].mean(axis=1)

# Show sample
print("\nTwitter sample with sentiment:")
display(twitter[['Entity','Text','clean_text','vader_score','textblob_polarity','sentiment_score']].head(5))

# ----------------------
# STEP 6: Aggregate sentiment per Entity and prepare for mapping
# ----------------------
sentiment_summary = twitter.groupby('Entity').agg(
    avg_vader = ('vader_score','mean'),
    avg_textblob = ('textblob_polarity','mean'),
    avg_sentiment = ('sentiment_score','mean'),
    mention_count = ('sentiment_score','count'),
    sentiment_std = ('sentiment_score','std')
).reset_index().rename(columns={'Entity':'entity'})

# Normalize entity key
sentiment_summary['entity_key'] = sentiment_summary['entity'].astype(str).str.lower().str.strip()

print("\nSentiment summary sample:")
display(sentiment_summary.head(6))

# ----------------------
# STEP 7: Prepare FIFA keys and basic features
# ----------------------
# normalize player names for matching
if 'name' in fifa.columns:
    fifa['name_key'] = fifa['name'].astype(str).str.lower().str.strip()
elif 'full_name' in fifa.columns:
    fifa['name_key'] = fifa['full_name'].astype(str).str.lower().str.strip()
else:
    # fallback to index
    fifa['name_key'] = fifa.index.astype(str)

# performance_score (example weighted aggregator); modify weights as you prefer
num_cols = fifa.select_dtypes(include=[np.number]).columns.tolist()
# prefer explicit columns if available
required = ['overall_rating','potential','stamina','acceleration']
if all(c in fifa.columns for c in required):
    fifa['performance_score'] = (
        fifa['overall_rating'] * 0.5 +
        fifa['potential'] * 0.3 +
        fifa['stamina'] * 0.1 +
        fifa['acceleration'] * 0.1
    )
else:
    fifa['performance_score'] = fifa[num_cols].mean(axis=1)

# rating trend (rolling on available chronological rating if present) — fallback to NaN
# If fifa contains year-wise rows per player, you'd compute trend; here we create a fallback rating_trend column
fifa['rating_trend'] = fifa['overall_rating'].rolling(window=5, min_periods=1).mean()

# ----------------------
# STEP 8: Injury features (aggregate per player)
# ----------------------
# Normalize injury key
if 'p_id2' in injury.columns:
    injury['p_id2_key'] = injury['p_id2'].astype(str).str.lower().str.strip()
else:
    # try other columns
    for c in ['player_id','player','name']:
        if c in injury.columns:
            injury.rename(columns={c:'p_id2'}, inplace=True)
            injury['p_id2_key'] = injury['p_id2'].astype(str).str.lower().str.strip()
            break

# compute injury impact metric
injury['total_days_injured'] = injury.get('total_days_injured', injury.get('season_days_injured', 0)).fillna(0)
injury['total_minutes_played'] = injury.get('total_minutes_played', injury.get('season_minutes_played', 0)).fillna(0)
injury['injury_impact'] = injury['total_days_injured'] / (injury['total_minutes_played'] + 1)

inj_agg = injury.groupby('p_id2_key').agg(
    total_days_injured = ('total_days_injured','sum'),
    avg_injury_impact = ('injury_impact','mean'),
    seasons_injured = ('season_days_injured','count')
).reset_index().rename(columns={'p_id2_key':'name_key'})

print("\nInjury aggregate sample:")
display(inj_agg.head(4))

# ----------------------
# STEP 9: Appearance features (per-player aggregates & simple form slope)
# ----------------------
# Normalize appearance names
if 'player_name' in appearance.columns:
    appearance['player_name_key'] = appearance['player_name'].astype(str).str.lower().str.strip()
else:
    appearance['player_name_key'] = appearance['player_id'].astype(str)

# ensure numeric columns exist; convert safely
for c in ['goals','minutes_played','assists','game_id','date']:
    if c in appearance.columns and appearance[c].dtype == 'O' and c!='date':
        appearance[c] = pd.to_numeric(appearance[c], errors='coerce')

# per-player aggregates
app_agg = appearance.groupby('player_name_key').agg(
    total_games = ('game_id','nunique') if 'game_id' in appearance.columns else ('appearance_id','count'),
    total_minutes = ('minutes_played','sum') if 'minutes_played' in appearance.columns else ('player_id','count'),
    total_goals = ('goals','sum') if 'goals' in appearance.columns else ('player_id','count'),
    total_assists = ('assists','sum') if 'assists' in appearance.columns else ('player_id','count')
).reset_index()

# derive per-90 metric and minutes per game
app_agg['minutes_per_game'] = app_agg['total_minutes'] / (app_agg['total_games'] + 1e-6)
app_agg['goals_per_90'] = app_agg['total_goals'] / (app_agg['total_minutes'] / 90 + 1e-6)

# form slope: compute slope of goals over time for players that have date info
if 'date' in appearance.columns:
    appearance['date_parsed'] = pd.to_datetime(appearance['date'], errors='coerce')
    def compute_slope(g):
        g = g.dropna(subset=['date_parsed','goals'])
        if len(g) < 3:
            return np.nan
        x = g['date_parsed'].map(pd.Timestamp.toordinal).values
        y = g['goals'].values
        try:
            slope = linregress(x, y)[0]
        except:
            slope = np.nan
        return slope
    slopes = appearance.groupby('player_name_key').apply(compute_slope).reset_index().rename(columns={0:'goals_slope'})
    app_agg = app_agg.merge(slopes, left_on='player_name_key', right_on='player_name_key', how='left')
else:
    app_agg['goals_slope'] = np.nan

print("\nAppearance aggregates sample:")
display(app_agg.head(4))

# ----------------------
# STEP 10: Valuation trends (time series per player)
# ----------------------
# ensure valuation has player id + date + market_value_in_eur
if 'player_id' in valuation.columns:
    valuation['player_id_key'] = valuation['player_id'].astype(str).str.lower().str.strip()
else:
    valuation['player_id_key'] = valuation.get('player', pd.Series()).astype(str).str.lower().str.strip()

if 'date' in valuation.columns:
    valuation['date_parsed'] = pd.to_datetime(valuation['date'], errors='coerce')

# compute average pct change and volatility per player
def valuation_trend(g):
    g = g.dropna(subset=['date_parsed','market_value_in_eur'])
    if len(g) < 2:
        return pd.Series({'avg_value_trend': np.nan, 'value_volatility': np.nan})
    g = g.sort_values('date_parsed')
    vals = g['market_value_in_eur'].astype(float).values
    pct = pd.Series(vals).pct_change().replace([np.inf,-np.inf], np.nan).dropna()
    avg_pct = pct.mean() if len(pct)>0 else 0.0
    vol = pct.std() if len(pct)>0 else 0.0
    # slope by linear regression
    x = g['date_parsed'].map(pd.Timestamp.toordinal).astype(float).values
    try:
        slope = linregress(x[~np.isnan(vals)], vals[~np.isnan(vals)])[0] if len(vals)>1 else np.nan
    except:
        slope = np.nan
    return pd.Series({'avg_value_trend': avg_pct, 'value_volatility': vol, 'value_slope': slope})

if not valuation.empty and 'market_value_in_eur' in valuation.columns:
    vt = valuation.groupby('player_id_key').apply(valuation_trend).reset_index()
else:
    vt = pd.DataFrame(columns=['player_id_key','avg_value_trend','value_volatility','value_slope'])

print("\nValuation trend sample:")
display(vt.head(4))

# ----------------------
# STEP 11: Map Twitter entities -> FIFA names using fuzzy matching
# ----------------------
# Create a list of fifa name keys to match to
fifa_names = fifa['name_key'].dropna().unique().tolist()

def best_match_entity_to_player(entity, choices, score_cutoff=75):
    if not isinstance(entity, str) or entity.strip()=='':
        return (None, 0)
    best = process.extractOne(entity, choices, scorer=fuzz.WRatio)
    if best is None:
        return (None, 0)
    name, score, idx = best
    if score >= score_cutoff:
        return (name, score)
    return (None, score)

# Apply matching for sentiment_summary
sentiment_summary['mapped_name'] = sentiment_summary['entity_key'].apply(
    lambda e: best_match_entity_to_player(e, fifa_names, score_cutoff=70)[0]
)
# For those not matched, mapped_name will be None
print("\nSample mapping results (entity -> mapped_name):")
display(sentiment_summary.head(6))

# ----------------------
# STEP 12: Prepare sentiment features keyed by fifa.name_key
# ----------------------
# If mapping exists, use mapped_name; otherwise attempt direct name equality
sentiment_summary['name_key'] = sentiment_summary['mapped_name'].fillna(sentiment_summary['entity_key'])
# keep necessary columns
sent_for_merge = sentiment_summary[['name_key','avg_vader','avg_textblob','avg_sentiment','mention_count','sentiment_std']].copy()
sent_for_merge.rename(columns={'avg_vader':'avg_vader_score','avg_textblob':'avg_textblob_score'}, inplace=True)
print("\nPrepared sentiment features sample:")
display(sent_for_merge.head(4))

# ----------------------
# STEP 13: Merge everything into final dataset
# ----------------------
# Merge order: fifa <- inj_agg <- app_agg <- sentiment <- valuation (value trends)
final = fifa.copy()
# merge injury on name_key
final = final.merge(inj_agg, left_on='name_key', right_on='name_key', how='left')
# merge appearance aggregates (player_name_key)
final = final.merge(app_agg, left_on='name_key', right_on='player_name_key', how='left')
# merge sentiment (name_key)
final = final.merge(sent_for_merge, left_on='name_key', right_on='name_key', how='left')
# merge valuation trends: attempt by player_id if available, otherwise try name_key
if 'player_id' in final.columns:
    final['player_id_key'] = final['player_id'].astype(str).str.lower().str.strip()
    final = final.merge(vt, left_on='player_id_key', right_on='player_id_key', how='left')
else:
    final = final.merge(vt, left_on='name_key', right_on='player_id_key', how='left')

# ----------------------
# STEP 14: Create combined features relevant to value modeling
# ----------------------
# sentiment-weighted metric
final['mention_count'] = final['mention_count'].fillna(0)
final['avg_sentiment'] = final['avg_sentiment'].fillna(0)
final['sentiment_weighted'] = final['avg_sentiment'] * (1 + np.log1p(final['mention_count']))

# injury-adjusted performance
final['avg_injury_impact'] = final['avg_injury_impact'].fillna(0)
final['performance_score'] = final['performance_score'].fillna(final.select_dtypes(include=[np.number]).mean(axis=1))
final['injury_adjusted_performance'] = final['performance_score'] * (1 - final['avg_injury_impact'])

# value per rating (if available)
if 'value_euro' in final.columns:
    final['value_per_rating'] = final['value_euro'] / (final['overall_rating'] + 1)

# fill numeric na with zeros for modeling convenience
numcols = final.select_dtypes(include=[np.number]).columns.tolist()
final[numcols] = final[numcols].fillna(0)

print("\nFinal merged feature set shape:", final.shape)
display(final.head(3))

# ----------------------
# STEP 15: Save final feature CSV
# ----------------------
OUT_PATH = '/content/final_feature_set.csv'
final.to_csv(OUT_PATH, index=False, encoding='utf-8-sig')
print("\nSaved final features to:", OUT_PATH)

# ----------------------
# STEP 16: Sentiment Analysis Report (text file + plots)
# ----------------------
report_lines = []
report_lines.append("Sentiment Analysis & Feature Engineering Report")
report_lines.append(f"Generated: {datetime.utcnow().isoformat()} UTC")
report_lines.append("")
report_lines.append("Top 20 entities by avg_sentiment (min mentions = 5):")
top_entities = sentiment_summary[sentiment_summary['mention_count']>=5].sort_values('avg_sentiment', ascending=False).head(20)
report_lines += top_entities[['entity','avg_sentiment','mention_count']].to_string(index=False).splitlines()
report_lines.append("")
report_lines.append("Bottom 20 entities by avg_sentiment (min mentions = 5):")
bottom_entities = sentiment_summary[sentiment_summary['mention_count']>=5].sort_values('avg_sentiment', ascending=True).head(20)
report_lines += bottom_entities[['entity','avg_sentiment','mention_count']].to_string(index=False).splitlines()
report_lines.append("")

# selected correlations
corr_cols = ['avg_sentiment','sentiment_weighted','injury_adjusted_performance','value_per_rating','value_euro','avg_value_trend']
present = [c for c in corr_cols if c in final.columns]
report_lines.append("Correlation matrix (selected features):")
if present:
    corr = final[present].corr()
    report_lines += corr.to_string().splitlines()
else:
    report_lines.append("Not enough columns to compute correlation.")

# Save report
REPORT_PATH = '/content/sentiment_feature_report.txt'
with open(REPORT_PATH,'w', encoding='utf-8') as f:
    f.write("\n".join(report_lines))

print("\nSaved report to:", REPORT_PATH)
print("\nReport preview:")
print("\n".join(report_lines[:40]))

# ----------------------
# STEP 17: Quick Visualizations
# (display them inline)
# ----------------------
# 1. Sentiment distribution (twitter)
plt.figure(figsize=(6,4))
sns.histplot(twitter['sentiment_score'], bins=30, kde=True)
plt.title("Distribution of Combined Sentiment Score (Twitter)")
plt.xlabel("Sentiment score")
plt.show()

# 2. Sentiment vs Market Value (if value_euro available)
if 'value_euro' in final.columns and final['value_euro'].sum() > 0:
    plt.figure(figsize=(7,5))
    sns.scatterplot(x='avg_sentiment', y='value_euro', size='mention_count', data=final, alpha=0.7)
    plt.title("Avg Sentiment vs Market Value (bubble size = mentions)")
    plt.xlabel("Avg Sentiment")
    plt.ylabel("Market Value (EUR)")
    plt.show()

# 3. Correlation heatmap
if len(present) >= 2:
    plt.figure(figsize=(7,6))
    sns.heatmap(final[present].corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title("Correlations among engineered features")
    plt.show()

# 4. Wordclouds (positive vs negative tweets)
positive_text = " ".join(twitter[twitter['sentiment_score']>0.1]['clean_text'].dropna().astype(str))
negative_text = " ".join(twitter[twitter['sentiment_score']<-0.1]['clean_text'].dropna().astype(str))

if len(positive_text) > 50:
    plt.figure(figsize=(10,5))
    wc = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
    plt.imshow(wc, interpolation='bilinear'); plt.axis('off'); plt.title('Positive Tweets WordCloud'); plt.show()
if len(negative_text) > 50:
    plt.figure(figsize=(10,5))
    wc2 = WordCloud(width=800, height=400, background_color='black', colormap='Reds').generate(negative_text)
    plt.imshow(wc2, interpolation='bilinear'); plt.axis('off'); plt.title('Negative Tweets WordCloud'); plt.show()

print("\nDone  — final features saved and report generated.")
```

```python
# ======================
# Advanced Feature Engineering + Sentiment Analysis
# Complete notebook code block — run in Colab / Jupyter
# ======================

# STEP A: Install required packages (Colab only)
# (If running locally, install these with pip if not already installed)
!pip install -q vaderSentiment textblob rapidfuzz wordcloud

# Download textblob corpora (if needed)
!python -m textblob.download_corpora >/dev/null 2>&1

# ----------------------
# STEP 1: Imports
# ----------------------
import pandas as pd
import numpy as np
import re
from datetime import datetime
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from rapidfuzz import process, fuzz
from scipy.stats import linregress
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
sns.set_style("whitegrid")

# ----------------------
# STEP 2: File paths (update if needed)
# ----------------------
# Use the cleaned filenames you mentioned earlier
FIFA_FILE = '/content/fifa_players_cleaned.csv'
INJURY_FILE = '/content/fifa_players_injury_cleaned.csv'
APPEAR_FILE = '/content/appearance_data_cleaned.csv'     # appearance dataset
VALUATION_FILE = '/content/player_valuations.csv'     # valuation timeseries
TW_TRAIN = '/content/twitter_training.csv'
TW_VAL = '/content/twitter_validation.csv'

# ----------------------
# STEP 3: Load datasets safely
# ----------------------
def safe_read(path, **kwargs):
    try:
        return pd.read_csv(path, **kwargs)
    except Exception as e:
        print(f"Warning: could not read {path}: {e}")
        return pd.DataFrame()

fifa = safe_read(FIFA_FILE, encoding='utf-8')
injury = safe_read(INJURY_FILE, encoding='utf-8')
appearance = safe_read(APPEAR_FILE, encoding='utf-8')
valuation = safe_read(VALUATION_FILE, encoding='utf-8')
tw_train = safe_read(TW_TRAIN, header=None, encoding='utf-8')  # headerless
tw_val = safe_read(TW_VAL, header=None, encoding='utf-8')

print("Shapes loaded — fifa:", fifa.shape, "injury:", injury.shape, "appearance:", appearance.shape, "valuation:", valuation.shape,
      "tw_train:", tw_train.shape, "tw_val:", tw_val.shape)

# ----------------------
# STEP 4: Prepare Twitter combined dataset (headerless -> assign columns)
# ----------------------
# Many of your twitter files were headerless and had 4 columns: ID, Entity, Sentiment, Text
def prepare_twitter(df):
    if df.empty:
        return df
    # if it already has column names like 'Text' or 'text', keep them
    cols = df.columns.tolist()
    # Typical case: headerless with 4 columns
    if len(cols) >= 4 and all(isinstance(c, (int, np.integer)) for c in cols[:4]):
        df = df.iloc[:, :4]  # take first 4 columns
        df.columns = ['ID', 'Entity', 'Label', 'Text']
    else:
        # try to standardize by finding a column likely to be text
        text_col = None
        for c in df.columns:
            if str(c).lower() in ['text', 'tweet', 'body']:
                text_col = c; break
        if text_col is None:
            # fallback: treat last object column as text
            obj_cols = df.select_dtypes(include='object').columns.tolist()
            if obj_cols:
                text_col = obj_cols[-1]
        # ensure at least Text column exists
        if text_col:
            df = df.rename(columns={text_col: 'Text'})
        # create entity if missing
        if 'Entity' not in df.columns:
            # try second column if exists
            if df.shape[1] >= 2:
                df = df.rename(columns={df.columns[1]: 'Entity'})
            else:
                df['Entity'] = 'unknown'
    return df

tw_train = prepare_twitter(tw_train)
tw_val = prepare_twitter(tw_val)
twitter = pd.concat([tw_train, tw_val], ignore_index=True)
twitter = twitter.loc[:, twitter.columns.drop_duplicates()]  # drop accidental dup cols
print("Twitter combined shape:", twitter.shape)
print("Twitter columns:", twitter.columns.tolist())
display(twitter.head(3))

# ----------------------
# STEP 5: Clean Twitter text and compute VADER + TextBlob scores
# ----------------------
def clean_text(s):
    s = str(s)
    s = s.replace('\n', ' ')
    s = re.sub(r'http\S+|www\.\S+', '', s)
    s = re.sub(r'@\w+', '', s)
    s = re.sub(r'#[A-Za-z0-9_]+', '', s)
    s = re.sub(r'[^A-Za-z0-9\s]', ' ', s)
    s = re.sub(r'\s+', ' ', s).strip().lower()
    return s

if 'Text' not in twitter.columns:
    twitter['Text'] = twitter.iloc[:, -1].astype(str)

twitter['clean_text'] = twitter['Text'].apply(clean_text)

# VADER + TextBlob
vader = SentimentIntensityAnalyzer()
twitter['vader_score'] = twitter['clean_text'].apply(lambda t: vader.polarity_scores(t)['compound'])
twitter['textblob_polarity'] = twitter['clean_text'].apply(lambda t: TextBlob(t).sentiment.polarity)
# combined robust score (mean)
twitter['sentiment_score'] = twitter[['vader_score','textblob_polarity']].mean(axis=1)

# Show sample
print("\nTwitter sample with sentiment:")
display(twitter[['Entity','Text','clean_text','vader_score','textblob_polarity','sentiment_score']].head(5))

# ----------------------
# STEP 6: Aggregate sentiment per Entity and prepare for mapping
# ----------------------
sentiment_summary = twitter.groupby('Entity').agg(
    avg_vader = ('vader_score','mean'),
    avg_textblob = ('textblob_polarity','mean'),
    avg_sentiment = ('sentiment_score','mean'),
    mention_count = ('sentiment_score','count'),
    sentiment_std = ('sentiment_score','std')
).reset_index().rename(columns={'Entity':'entity'})

# Normalize entity key
sentiment_summary['entity_key'] = sentiment_summary['entity'].astype(str).str.lower().str.strip()

print("\nSentiment summary sample:")
display(sentiment_summary.head(6))

# ----------------------
# STEP 7: Prepare FIFA keys and basic features
# ----------------------
# normalize player names for matching
if 'name' in fifa.columns:
    fifa['name_key'] = fifa['name'].astype(str).str.lower().str.strip()
elif 'full_name' in fifa.columns:
    fifa['name_key'] = fifa['full_name'].astype(str).str.lower().str.strip()
else:
    # fallback to index
    fifa['name_key'] = fifa.index.astype(str)

# performance_score (example weighted aggregator); modify weights as you prefer
num_cols = fifa.select_dtypes(include=[np.number]).columns.tolist()
# prefer explicit columns if available
required = ['overall_rating','potential','stamina','acceleration']
if all(c in fifa.columns for c in required):
    fifa['performance_score'] = (
        fifa['overall_rating'] * 0.5 +
        fifa['potential'] * 0.3 +
        fifa['stamina'] * 0.1 +
        fifa['acceleration'] * 0.1
    )
else:
    fifa['performance_score'] = fifa[num_cols].mean(axis=1)

# rating trend (rolling on available chronological rating if present) — fallback to NaN
# If fifa contains year-wise rows per player, you'd compute trend; here we create a fallback rating_trend column
fifa['rating_trend'] = fifa['overall_rating'].rolling(window=5, min_periods=1).mean()

# ----------------------
# STEP 8: Injury features (aggregate per player)
# ----------------------
# Normalize injury key
if 'p_id2' in injury.columns:
    injury['p_id2_key'] = injury['p_id2'].astype(str).str.lower().str.strip()
else:
    # try other columns
    for c in ['player_id','player','name']:
        if c in injury.columns:
            injury.rename(columns={c:'p_id2'}, inplace=True)
            injury['p_id2_key'] = injury['p_id2'].astype(str).str.lower().str.strip()
            break

# compute injury impact metric
injury['total_days_injured'] = injury.get('total_days_injured', injury.get('season_days_injured', 0)).fillna(0)
injury['total_minutes_played'] = injury.get('total_minutes_played', injury.get('season_minutes_played', 0)).fillna(0)
injury['injury_impact'] = injury['total_days_injured'] / (injury['total_minutes_played'] + 1)

inj_agg = injury.groupby('p_id2_key').agg(
    total_days_injured = ('total_days_injured','sum'),
    avg_injury_impact = ('injury_impact','mean'),
    seasons_injured = ('season_days_injured','count')
).reset_index().rename(columns={'p_id2_key':'name_key'})

print("\nInjury aggregate sample:")
display(inj_agg.head(4))

# ----------------------
# STEP 9: Appearance features (per-player aggregates & simple form slope)
# ----------------------
# Normalize appearance names
if 'player_name' in appearance.columns:
    appearance['player_name_key'] = appearance['player_name'].astype(str).str.lower().str.strip()
else:
    appearance['player_name_key'] = appearance['player_id'].astype(str)

# ensure numeric columns exist; convert safely
for c in ['goals','minutes_played','assists','game_id','date']:
    if c in appearance.columns and appearance[c].dtype == 'O' and c!='date':
        appearance[c] = pd.to_numeric(appearance[c], errors='coerce')

# per-player aggregates
app_agg = appearance.groupby('player_name_key').agg(
    total_games = ('game_id','nunique') if 'game_id' in appearance.columns else ('appearance_id','count'),
    total_minutes = ('minutes_played','sum') if 'minutes_played' in appearance.columns else ('player_id','count'),
    total_goals = ('goals','sum') if 'goals' in appearance.columns else ('player_id','count'),
    total_assists = ('assists','sum') if 'assists' in appearance.columns else ('player_id','count')
).reset_index()

# derive per-90 metric and minutes per game
app_agg['minutes_per_game'] = app_agg['total_minutes'] / (app_agg['total_games'] + 1e-6)
app_agg['goals_per_90'] = app_agg['total_goals'] / (app_agg['total_minutes'] / 90 + 1e-6)

# form slope: compute slope of goals over time for players that have date info
if 'date' in appearance.columns:
    appearance['date_parsed'] = pd.to_datetime(appearance['date'], errors='coerce')
    def compute_slope(g):
        g = g.dropna(subset=['date_parsed','goals'])
        if len(g) < 3:
            return np.nan
        x = g['date_parsed'].map(pd.Timestamp.toordinal).values
        y = g['goals'].values
        try:
            slope = linregress(x, y)[0]
        except:
            slope = np.nan
        return slope
    slopes = appearance.groupby('player_name_key').apply(compute_slope).reset_index().rename(columns={0:'goals_slope'})
    app_agg = app_agg.merge(slopes, left_on='player_name_key', right_on='player_name_key', how='left')
else:
    app_agg['goals_slope'] = np.nan

print("\nAppearance aggregates sample:")
display(app_agg.head(4))

# ----------------------
# STEP 10: Valuation trends (time series per player)
# ----------------------
# ensure valuation has player id + date + market_value_in_eur
if 'player_id' in valuation.columns:
    valuation['player_id_key'] = valuation['player_id'].astype(str).str.lower().str.strip()
else:
    valuation['player_id_key'] = valuation.get('player', pd.Series()).astype(str).str.lower().str.strip()

if 'date' in valuation.columns:
    valuation['date_parsed'] = pd.to_datetime(valuation['date'], errors='coerce')

# compute average pct change and volatility per player
def valuation_trend(g):
    g = g.dropna(subset=['date_parsed','market_value_in_eur'])
    if len(g) < 2:
        return pd.Series({'avg_value_trend': np.nan, 'value_volatility': np.nan})
    g = g.sort_values('date_parsed')
    vals = g['market_value_in_eur'].astype(float).values
    pct = pd.Series(vals).pct_change().replace([np.inf,-np.inf], np.nan).dropna()
    avg_pct = pct.mean() if len(pct)>0 else 0.0
    vol = pct.std() if len(pct)>0 else 0.0
    # slope by linear regression
    x = g['date_parsed'].map(pd.Timestamp.toordinal).astype(float).values
    try:
        slope = linregress(x[~np.isnan(vals)], vals[~np.isnan(vals)])[0] if len(vals)>1 else np.nan
    except:
        slope = np.nan
    return pd.Series({'avg_value_trend': avg_pct, 'value_volatility': vol, 'value_slope': slope})

if not valuation.empty and 'market_value_in_eur' in valuation.columns:
    vt = valuation.groupby('player_id_key').apply(valuation_trend).reset_index()
else:
    vt = pd.DataFrame(columns=['player_id_key','avg_value_trend','value_volatility','value_slope'])

print("\nValuation trend sample:")
display(vt.head(4))

# ----------------------
# STEP 11: Map Twitter entities -> FIFA names using fuzzy matching
# ----------------------
# Create a list of fifa name keys to match to
fifa_names = fifa['name_key'].dropna().unique().tolist()

def best_match_entity_to_player(entity, choices, score_cutoff=75):
    if not isinstance(entity, str) or entity.strip()=='':
        return (None, 0)
    best = process.extractOne(entity, choices, scorer=fuzz.WRatio)
    if best is None:
        return (None, 0)
    name, score, idx = best
    if score >= score_cutoff:
        return (name, score)
    return (None, score)

# Apply matching for sentiment_summary
sentiment_summary['mapped_name'] = sentiment_summary['entity_key'].apply(
    lambda e: best_match_entity_to_player(e, fifa_names, score_cutoff=70)[0]
)
# For those not matched, mapped_name will be None
print("\nSample mapping results (entity -> mapped_name):")
display(sentiment_summary.head(6))

# ----------------------
# STEP 12: Prepare sentiment features keyed by fifa.name_key
# ----------------------
# If mapping exists, use mapped_name; otherwise attempt direct name equality
sentiment_summary['name_key'] = sentiment_summary['mapped_name'].fillna(sentiment_summary['entity_key'])
# keep necessary columns
sent_for_merge = sentiment_summary[['name_key','avg_vader','avg_textblob','avg_sentiment','mention_count','sentiment_std']].copy()
sent_for_merge.rename(columns={'avg_vader':'avg_vader_score','avg_textblob':'avg_textblob_score'}, inplace=True)
print("\nPrepared sentiment features sample:")
display(sent_for_merge.head(4))

# ----------------------
# STEP 13: Merge everything into final dataset
# ----------------------
# Merge order: fifa <- inj_agg <- app_agg <- sentiment <- valuation (value trends)
final = fifa.copy()
# merge injury on name_key
final = final.merge(inj_agg, left_on='name_key', right_on='name_key', how='left')
# merge appearance aggregates (player_name_key)
final = final.merge(app_agg, left_on='name_key', right_on='player_name_key', how='left')
# merge sentiment (name_key)
final = final.merge(sent_for_merge, left_on='name_key', right_on='name_key', how='left')
# merge valuation trends: attempt by player_id if available, otherwise try name_key
if 'player_id' in final.columns:
    final['player_id_key'] = final['player_id'].astype(str).str.lower().str.strip()
    final = final.merge(vt, left_on='player_id_key', right_on='player_id_key', how='left')
else:
    final = final.merge(vt, left_on='name_key', right_on='player_id_key', how='left')

# ----------------------
# STEP 14: Create combined features relevant to value modeling
# ----------------------
# sentiment-weighted metric
final['mention_count'] = final['mention_count'].fillna(0)
final['avg_sentiment'] = final['avg_sentiment'].fillna(0)
final['sentiment_weighted'] = final['avg_sentiment'] * (1 + np.log1p(final['mention_count']))

# injury-adjusted performance
final['avg_injury_impact'] = final['avg_injury_impact'].fillna(0)
final['performance_score'] = final['performance_score'].fillna(final.select_dtypes(include=[np.number]).mean(axis=1))
final['injury_adjusted_performance'] = final['performance_score'] * (1 - final['avg_injury_impact'])

# value per rating (if available)
if 'value_euro' in final.columns:
    final['value_per_rating'] = final['value_euro'] / (final['overall_rating'] + 1)

# fill numeric na with zeros for modeling convenience
numcols = final.select_dtypes(include=[np.number]).columns.tolist()
final[numcols] = final[numcols].fillna(0)

print("\nFinal merged feature set shape:", final.shape)
display(final.head(3))

# ----------------------
# STEP 15: Save final feature CSV
# ----------------------
OUT_PATH = '/content/final_feature_set.csv'
final.to_csv(OUT_PATH, index=False, encoding='utf-8-sig')
print("\nSaved final features to:", OUT_PATH)

# ----------------------
# STEP 16: Sentiment Analysis Report (text file + plots)
# ----------------------
report_lines = []
report_lines.append("Sentiment Analysis & Feature Engineering Report")
report_lines.append(f"Generated: {datetime.utcnow().isoformat()} UTC")
report_lines.append("")
report_lines.append("Top 20 entities by avg_sentiment (min mentions = 5):")
top_entities = sentiment_summary[sentiment_summary['mention_count']>=5].sort_values('avg_sentiment', ascending=False).head(20)
report_lines += top_entities[['entity','avg_sentiment','mention_count']].to_string(index=False).splitlines()
report_lines.append("")
report_lines.append("Bottom 20 entities by avg_sentiment (min mentions = 5):")
bottom_entities = sentiment_summary[sentiment_summary['mention_count']>=5].sort_values('avg_sentiment', ascending=True).head(20)
report_lines += bottom_entities[['entity','avg_sentiment','mention_count']].to_string(index=False).splitlines()
report_lines.append("")

# selected correlations
corr_cols = ['avg_sentiment','sentiment_weighted','injury_adjusted_performance','value_per_rating','value_euro','avg_value_trend']
present = [c for c in corr_cols if c in final.columns]
report_lines.append("Correlation matrix (selected features):")
if present:
    corr = final[present].corr()
    report_lines += corr.to_string().splitlines()
else:
    report_lines.append("Not enough columns to compute correlation.")

# Save report
REPORT_PATH = '/content/sentiment_feature_report.txt'
with open(REPORT_PATH,'w', encoding='utf-8') as f:
    f.write("\n".join(report_lines))

print("\nSaved report to:", REPORT_PATH)
print("\nReport preview:")
print("\n".join(report_lines[:40]))

# ----------------------
# STEP 17: Quick Visualizations
# (display them inline)
# ----------------------
# 1. Sentiment distribution (twitter)
plt.figure(figsize=(6,4))
sns.histplot(twitter['sentiment_score'], bins=30, kde=True)
plt.title("Distribution of Combined Sentiment Score (Twitter)")
plt.xlabel("Sentiment score")
plt.show()

# 2. Sentiment vs Market Value (if value_euro available)
if 'value_euro' in final.columns and final['value_euro'].sum() > 0:
    plt.figure(figsize=(7,5))
    sns.scatterplot(x='avg_sentiment', y='value_euro', size='mention_count', data=final, alpha=0.7)
    plt.title("Avg Sentiment vs Market Value (bubble size = mentions)")
    plt.xlabel("Avg Sentiment")
    plt.ylabel("Market Value (EUR)")
    plt.show()

# 3. Correlation heatmap
if len(present) >= 2:
    plt.figure(figsize=(7,6))
    sns.heatmap(final[present].corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title("Correlations among engineered features")
    plt.show()

# 4. Wordclouds (positive vs negative tweets)
positive_text = " ".join(twitter[twitter['sentiment_score']>0.1]['clean_text'].dropna().astype(str))
negative_text = " ".join(twitter[twitter['sentiment_score']<-0.1]['clean_text'].dropna().astype(str))

if len(positive_text) > 50:
    plt.figure(figsize=(10,5))
    wc = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
    plt.imshow(wc, interpolation='bilinear'); plt.axis('off'); plt.title('Positive Tweets WordCloud'); plt.show()
if len(negative_text) > 50:
    plt.figure(figsize=(10,5))
    wc2 = WordCloud(width=800, height=400, background_color='black', colormap='Reds').generate(negative_text)
    plt.imshow(wc2, interpolation='bilinear'); plt.axis('off'); plt.title('Negative Tweets WordCloud'); plt.show()

print("\nDone  — final features saved and report generated.")
```
