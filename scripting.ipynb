```python
import pandas as pd

inputfile = 'dataset.csv'
outputfile = 'results/injury_cleaned.csv'
df = pd.read_csv(inputfile)
print("Rows before cleaning:", len(df))
df = df.drop_duplicates()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna('Unknown', inplace=True)
    else:
        df[col].fillna(0, inplace=True)
numericcols = [
    'start_year', 'season_days_injured', 'total_days_injured', 'season_minutes_played',
    'season_games_played', 'season_matches_in_squad', 'total_minutes_played', 'total_games_played',
    'height_cm', 'weight_kg', 'age'
]
for col in numericcols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
df.to_csv(outputfile, index=False)
print("Cleaned injury dataset saved.")
print("Rows after cleaning:", len(df))
```

```python
import pandas as pd
import os

datafolder = 'archive (2)/data'  # Corrected path with space and (2)
outputfolder = 'results/'
os.makedirs(outputfolder, exist_ok=True)

files = ['competitions.csv', 'matches.csv', 'events.csv', 'lineups.csv', 'three-sixty.csv']

for filename in files:
    inputfile = os.path.join(datafolder, filename)
    outputfile = os.path.join(outputfolder, f"{filename.split('.')[0]}_cleaned.csv")
    if not os.path.exists(inputfile):
        print(f"{filename} not found, skipping...")
        continue
    print(f"Cleaning {filename}...")
    try:
        chunklist = []
        for chunk in pd.read_csv(inputfile, chunksize=500000):
            chunk = chunk.drop_duplicates()
            for col in chunk.columns:
                if chunk[col].dtype == 'object':
                    chunk[col].fillna('Unknown', inplace=True)
                else:
                    chunk[col].fillna(0, inplace=True)
                if chunk[col].dtype in ['int64', 'float64']:
                    chunk[col] = pd.to_numeric(chunk[col], errors='coerce').fillna(0)
            chunklist.append(chunk)
        df = pd.concat(chunklist, ignore_index=True)
        df.to_csv(outputfile, index=False)
        print(f"{filename} cleaned. Rows after cleaning: {len(df)}")
    except Exception as e:
        print(f"Error cleaning {filename}: {e}")
```

```python
import os
import json
import pandas as pd
```

```python

# Set your base directory where the data is extracted
BASE_DIR = 'C:/Users/akshi/Downloads/dyyyyna/stackbombs/'

def load_json_files_to_df(folder, limit=None):
    all_data = []
    count = 0
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith('.json'):
                with open(os.path.join(root, file), encoding='utf-8') as f:
                    data = json.load(f)
                    if isinstance(data, dict):
                        all_data.append(data)
                    elif isinstance(data, list):
                        all_data.extend(data)
                count += 1
                if limit and count >= limit:
                    break
        if limit and count >= limit:
            break
    df = pd.json_normalize(all_data, sep='_')
    return df


def drop_duplicates_safe(df):
    list_cols = [col for col in df.columns if df[col].apply(lambda x: isinstance(x, list)).any()]
    df_no_list = df.drop(columns=list_cols)
    df_no_list = df_no_list.drop_duplicates()
    df_cleaned = pd.concat([df_no_list, df[list_cols]], axis=1)
    return df_cleaned


# Load competitions
competitions_path = os.path.join(BASE_DIR, 'data', 'competitions.json')
competitions = pd.read_json(competitions_path)
competitions.drop_duplicates(inplace=True)
competitions.dropna(axis=1, how='all', inplace=True)

# Load matches data
matches_path = os.path.join(BASE_DIR, 'data', 'matches')
matches_df = load_json_files_to_df(matches_path)
matches_df = drop_duplicates_safe(matches_df)
matches_df.dropna(axis=1, how='all', inplace=True)

# Load events data (You can limit for faster testing)
events_path = os.path.join(BASE_DIR, 'data', 'events')
events_df = load_json_files_to_df(events_path, limit=20)  # set limit=None to load all
events_df = drop_duplicates_safe(events_df)
events_df.dropna(axis=1, how='all', inplace=True)

# Load lineups data
lineups_path = os.path.join(BASE_DIR, 'data', 'lineups')
lineups_df = load_json_files_to_df(lineups_path)
lineups_df = drop_duplicates_safe(lineups_df)
lineups_df.dropna(axis=1, how='all', inplace=True)

# Remove columns with >90% missing data
def remove_high_na_columns(df, threshold=0.9):
    na_frac = df.isna().mean()
    return df.loc[:, na_frac < threshold]

competitions = remove_high_na_columns(competitions)
matches_df = remove_high_na_columns(matches_df)
events_df = remove_high_na_columns(events_df)
lineups_df = remove_high_na_columns(lineups_df)

# Save cleaned CSV files
competitions.to_csv(os.path.join(BASE_DIR, 'cleaned_competitions.csv'), index=False)
matches_df.to_csv(os.path.join(BASE_DIR, 'cleaned_matches.csv'), index=False)
events_df.to_csv(os.path.join(BASE_DIR, 'cleaned_events.csv'), index=False)
lineups_df.to_csv(os.path.join(BASE_DIR, 'cleaned_lineups.csv'), index=False)

print("Cleaning complete. Cleaned CSV files saved.")
```

```python
# Twitter API v2 Authentication (robust)
# This cell will try to use environment variables first. If you prefer hardcoded values, set them below
import os

# Prefer environment variables for security
bearer = os.environ.get('TWITTER_BEARER_TOKEN')
api_key = os.environ.get('TWITTER_API_KEY')
api_secret = os.environ.get('TWITTER_API_SECRET')
access_token = os.environ.get('TWITTER_ACCESS_TOKEN')
access_secret = os.environ.get('TWITTER_ACCESS_SECRET')

# Fallback: if you previously hard-coded values in this notebook, you can uncomment and set them here.
# bearer = bearer or "<PUT_BEARER_TOKEN_HERE>"
# api_key = api_key or "<PUT_API_KEY_HERE>"
# api_secret = api_secret or "<PUT_API_SECRET_HERE>"
# access_token = access_token or "<PUT_ACCESS_TOKEN_HERE>"
# access_secret = access_secret or "<PUT_ACCESS_SECRET_HERE>"

# Build credentials dict only with values available
creds = {}
if bearer:
    creds['bearer_token'] = bearer
if api_key:
    creds['consumer_key'] = api_key
if api_secret:
    creds['consumer_secret'] = api_secret
if access_token:
    creds['access_token'] = access_token
if access_secret:
    creds['access_token_secret'] = access_secret

client = None
if not creds:
    print("No Twitter credentials found in environment. Set TWITTER_BEARER_TOKEN or the other TWITTER_* env vars.")
else:
    try:
        client = tweepy.Client(**creds, wait_on_rate_limit=True)
        # quick sanity call (may be None if permissions limited)
        try:
            me = client.get_me()
            if me and getattr(me, 'data', None):
                print(f"Authenticated as user id={me.data.id}")
            else:
                print("Client created but could not fetch user info (this can be normal if credentials are limited).")
        except Exception:
            # don't fail if get_me is not available for the keys used
            print("Client created (skipped get_me check).")
    except Exception as e:
        client = None
        print("Failed to create tweepy.Client:", str(e))

# After running this cell, ensure `client` is not None before calling API methods.
```
